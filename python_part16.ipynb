## Pandas Introduction

### What is Pandas?

- Pandas is a Python library for data analysis.
- It is used to work with tabular data.
- It helps in reading, cleaning, filtering, sorting, and analyzing data.

### Why Use Pandas?

- Read CSV files.
- Read Excel files.
- Clean missing data.
- Analyze large datasets.
- Perform statistical operations.

## Pandas Series

### What is a Series?

- A Series is a one-dimensional (1D) data structure.
- It stores a single column of data.
- Each value has an index (label).

### Create a Series

```python
pd.Series(data)
```

### Default Index

- Starts from 0.

### Custom Index

```python
pd.Series(data, index=[...])
```

### Access Values

```python
series[0]
series["Ali"]
```

### Create Series from Dictionary

```python
pd.Series(dictionary)
```

### Best Practice

- Use labels when data has meaningful names.
- Use Series for single-column data.

````python
import pandas as pd

# ==========================================
# Create a Series
# ==========================================

marks = pd.Series([80, 90, 70, 95])

print("Simple Series:")
print(marks)

# ==========================================
# Access by Index
# ==========================================

print("\nValue at Index 2:")
print(marks[2])

# ==========================================
# Custom Index
# ==========================================

marks = pd.Series(
    [80, 90, 70],
    index=["Ali", "Sara", "Ahmed"]
)

print("\nSeries with Custom Labels:")
print(marks)

print("\nMarks of Sara:")
print(marks["Sara"])

# ==========================================
# Series from Dictionary
# ==========================================

student = {
    "Ali": 80,
    "Sara": 90,
    "Ahmed": 70
}

series = pd.Series(student)

print("\nSeries from Dictionary:")
print(series)

# ==========================================
# Select Specific Labels
# ==========================================

selected = pd.Series(student, index=["Ali", "Ahmed"])

print("\nSelected Labels:")
print(selected)
````

## Pandas DataFrames

### What is a DataFrame?

- A DataFrame is a two-dimensional (2D) data structure.
- It contains rows and columns.
- Similar to an Excel spreadsheet or SQL table.

### Create a DataFrame

```python
pd.DataFrame(data)
```

### Data Source

- Dictionary keys become column names.
- Dictionary values become column data.

### Custom Row Labels

```python
pd.DataFrame(data, index=["A", "B", "C"])
```

### Access Rows

```python
df.loc["A"]
df.loc[["A", "C"]]
```

### Best Practice

- Store tabular data in DataFrames.
- Use meaningful column names.
- Use `loc` to access rows by label.

````python
import pandas as pd

# ==========================================
# Create a DataFrame
# ==========================================

data = {
    "Name": ["Ali", "Sara", "Ahmed"],
    "Age": [20, 21, 19],
    "City": ["Lahore", "Karachi", "Islamabad"]
}

df = pd.DataFrame(data)

print("Simple DataFrame:")
print(df)

# ==========================================
# DataFrame with Custom Index
# ==========================================

df2 = pd.DataFrame(
    data,
    index=["A", "B", "C"]
)

print("\nDataFrame with Custom Index:")
print(df2)

# ==========================================
# Access One Row
# ==========================================

print("\nRow B:")
print(df2.loc["B"])

# ==========================================
# Access Multiple Rows
# ==========================================

print("\nRows A and C:")
print(df2.loc[["A", "C"]])
````

## Pandas - Cleaning Data

### What is Data Cleaning?

- Data Cleaning means fixing or removing incorrect data.
- It prepares data for analysis and machine learning.

### Common Problems

- Missing values (NaN)
- Empty cells
- Wrong data
- Duplicate rows
- Wrong data format

### Remove Missing Rows

```python
df.dropna()
```

### Fill Missing Values

```python
df.fillna(0)
```

### Remove Duplicate Rows

```python
df.drop_duplicates()
```

### Update Wrong Data

```python
df.loc[row, "Column"] = value
```

### Best Practice

1. Load data.
2. Check dataset using `info()`.
3. Handle missing values.
4. Remove duplicates.
5. Continue with analysis.

````python
import pandas as pd

# ==========================================
# Read Dataset
# ==========================================

df = pd.read_csv("heart.csv")

# ==========================================
# Check Dataset Information
# ==========================================

print("Dataset Information:")
df.info()

# ==========================================
# Check Missing Values
# ==========================================

print("\nMissing Values:")
print(df.isnull().sum())

# ==========================================
# Remove Rows with Missing Values
# ==========================================

clean_df = df.dropna()

print("\nAfter Removing Missing Values:")
print(clean_df)

# ==========================================
# Fill Missing Values with 0
# ==========================================

filled_df = df.fillna(0)

print("\nAfter Filling Missing Values:")
print(filled_df)

# ==========================================
# Remove Duplicate Rows
# ==========================================

unique_df = df.drop_duplicates()

print("\nAfter Removing Duplicates:")
print(unique_df)
````

## Pandas - Cleaning Empty Cells

### Empty Cells

- Missing values are shown as NaN.
- NaN means Not a Number.

### Check Missing Values

```python
df.isnull().sum()
```

### Remove Missing Rows

```python
df.dropna()
```

### Permanently Remove

```python
df.dropna(inplace=True)
```

### Fill Missing Values

```python
df.fillna(0)
```

### Fill Specific Column

```python
df["Age"] = df["Age"].fillna(25)
```

### Fill with Mean

```python
x = df["Age"].mean()
df["Age"] = df["Age"].fillna(x)
```

### Fill with Median

```python
x = df["Age"].median()
df["Age"] = df["Age"].fillna(x)
```

### Fill with Mode

```python
x = df["Age"].mode()[0]
df["Age"] = df["Age"].fillna(x)
```

### Best Practice

- Check missing values first.
- Don't always use `0`.
- Choose Mean, Median or Mode according to the data.

````python
import pandas as pd

# Read Dataset
df = pd.read_csv("heart.csv")

# ==========================================
# Check Missing Values
# ==========================================

print("Missing Values:")
print(df.isnull().sum())

# ==========================================
# Remove Missing Rows
# ==========================================

print("\nAfter dropna():")
print(df.dropna())

# ==========================================
# Fill All Missing Values with 0
# ==========================================

print("\nAfter fillna(0):")
print(df.fillna(0))

# ==========================================
# Fill Age Column with Mean
# ==========================================

mean_age = df["Age"].mean()
df["Age"] = df["Age"].fillna(mean_age)

print("\nAfter Filling Age with Mean:")
print(df)

# ==========================================
# Fill Age Column with Median
# ==========================================

median_age = df["Age"].median()
df["Age"] = df["Age"].fillna(median_age)

print("\nAfter Filling Age with Median:")
print(df)

# ==========================================
# Fill Age Column with Mode
# ==========================================

mode_age = df["Age"].mode()[0]
df["Age"] = df["Age"].fillna(mode_age)

print("\nAfter Filling Age with Mode:")
print(df)
````

## Pandas - Cleaning Wrong Data

### Wrong Data

- Wrong data means incorrect or unrealistic values.
- Example: Age = 250

### Update a Value

```python
df.loc[1, "Age"] = 25
```

### Update Multiple Values

```python
df.loc[2, "Age"] = 30
df.loc[4, "Age"] = 28
```

### Fix Wrong Values Using Loop

```python
for i in df.index:
    if df.loc[i, "Age"] > 120:
        df.loc[i, "Age"] = 120
```

### Remove Wrong Rows

```python
for i in df.index:
    if df.loc[i, "Age"] > 120:
        df.drop(i, inplace=True)
```

### Best Practice

- Update the value if you know the correct value.
- Remove the row if the value is completely invalid.

````python
import pandas as pd

# Sample Data
data = {
    "Name": ["Ali", "Sara", "Ahmed", "Ayesha"],
    "Age": [20, 250, 19, 150]
}

df = pd.DataFrame(data)

print("Original Data:")
print(df)

# ==========================================
# Update Wrong Value
# ==========================================

df.loc[1, "Age"] = 25

print("\nAfter Updating Sara's Age:")
print(df)

# ==========================================
# Replace Ages Greater Than 120
# ==========================================

for i in df.index:
    if df.loc[i, "Age"] > 120:
        df.loc[i, "Age"] = 120

print("\nAfter Fixing Wrong Ages:")
print(df)

# ==========================================
# Remove Rows with Wrong Data
# ==========================================

df = pd.DataFrame(data)

for i in df.index:
    if df.loc[i, "Age"] > 120:
        df.drop(i, inplace=True)

print("\nAfter Removing Wrong Rows:")
print(df)
````

## Pandas - Removing Duplicates

### Duplicate Rows

- Duplicate rows contain the same data more than once.

### Check Duplicates

```python
df.duplicated()
```

- False → Unique row
- True → Duplicate row

### Count Duplicates

```python
df.duplicated().sum()
```

### Remove Duplicates

```python
df.drop_duplicates()
```

### Permanently Remove

```python
df.drop_duplicates(inplace=True)
```

### Best Practice

- Check duplicates before removing them.
- Use `inplace=True` if you want to update the original DataFrame.

````python
import pandas as pd

# Sample Data
data = {
    "Name": ["Ali", "Sara", "Ali", "Ahmed", "Sara"],
    "Age": [20, 21, 20, 19, 21],
    "City": ["Lahore", "Karachi", "Lahore", "Islamabad", "Karachi"]
}

df = pd.DataFrame(data)

# ==========================================
# Original Data
# ==========================================

print("Original Data:")
print(df)

# ==========================================
# Check Duplicate Rows
# ==========================================

print("\nDuplicate Rows:")
print(df.duplicated())

# ==========================================
# Count Duplicate Rows
# ==========================================

print("\nTotal Duplicates:")
print(df.duplicated().sum())

# ==========================================
# Remove Duplicate Rows
# ==========================================

new_df = df.drop_duplicates()

print("\nAfter Removing Duplicates:")
print(new_df)

# ==========================================
# Permanently Remove Duplicates
# ==========================================

df.drop_duplicates(inplace=True)

print("\nOriginal Data After Removing Duplicates:")
print(df)
````

## Pandas - Correlations

### What is Correlation?

- Correlation shows the relationship between two numeric columns.
- Values range from -1 to +1.

### Find Correlation

```python
df.corr()
```

### Correlation Values

- 1 → Perfect Positive
- 0 → No Correlation
- -1 → Perfect Negative

### Positive Correlation

- One value increases.
- The other value also increases.

Example:

- Study Hours
- Marks

### Negative Correlation

- One value increases.
- The other decreases.

Example:

- Exercise
- Body Fat

### No Correlation

- No meaningful relationship between columns.

### Best Practice

- Use `corr()` only for numeric columns.
- High correlation helps identify related features.

````python
import pandas as pd

df = pd.read_csv("data.csv")

print(df.corr())
````
