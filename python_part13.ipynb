## NumPy Array Indexing

### What is Indexing?

- Indexing is used to access specific elements in an array.
- NumPy indexing starts from `0`.

### 1D Indexing

Syntax:

```python
arr[index]
```

Example:

```python
arr = np.array([10,20,30])

print(arr[0])
```

### Negative Indexing

- Negative indexing starts from the end.

Example:

```python
arr[-1]
```

returns the last element.

### 2D Indexing

Syntax:

```python
arr[row, column]
```

Example:

```python
arr[1,2]
```

### 3D Indexing

Syntax:

```python
arr[layer, row, column]
```

Example:

```python
arr[0,1,1]
```

### Best Practice

- Remember indexing starts from `0`.
- For 2D arrays, always think **Row → Column**.
- For 3D arrays, think **Layer → Row → Column**.

````python
import numpy as np


# 1D Indexing


arr1 = np.array([10, 20, 30, 40])

print("First Element :", arr1[0])
print("Second Element:", arr1[1])
print("Last Element  :", arr1[-1])

print("----------------------")


# Sum of Elements

arr = np.array([1, 2, 3, 4])

print("Sum:", arr[2] + arr[3])

print("----------------------")


# 2D Indexing

arr2 = np.array([
    [10, 20, 30],
    [40, 50, 60]
])

print(arr2)

print("Row 0, Column 1:", arr2[0, 1])
print("Row 1, Column 2:", arr2[1, 2])

print("----------------------")


# 3D Indexing

arr3 = np.array([
    [
        [1, 2],
        [3, 4]
    ],
    [
        [5, 6],
        [7, 8]
    ]
])

print(arr3)

print("Layer 0, Row 1, Column 1:", arr3[0, 1, 1])
print("Layer 1, Row 0, Column 1:", arr3[1, 0, 1])
````

## NumPy Array Slicing

### What is Slicing?

- Slicing is used to access a portion of an array.
- It returns multiple elements.

### Basic Syntax

```python
arr[start:stop]
```

- Start index is included.
- Stop index is excluded.

### Examples

```python
arr[1:4]
```

```python
arr[:3]
```

```python
arr[2:]
```

```python
arr[:]
```

### Negative Slicing

```python
arr[-3:-1]
```

### Step

Syntax:

```python
arr[start:stop:step]
```

Example:

```python
arr[::2]
```

### Reverse Array

```python
arr[::-1]
```

### 2D Slicing

Single row:

```python
arr[0,1:3]
```

All rows:

```python
arr[:,1]
```

Multiple columns:

```python
arr[:,1:3]
```

### Best Practice

- Remember:
  - Start is included.
  - Stop is excluded.
- Use `:` to select all rows or all columns.

````python
import numpy as np

# ==========================================
# Create 1D Array
# ==========================================

arr = np.array([10, 20, 30, 40, 50])

print("Original Array:")
print(arr)

print("----------------------")

# Slice from index 1 to 3
print(arr[1:4])

# From beginning to index 2
print(arr[:3])

# From index 2 to end
print(arr[2:])

# Complete array
print(arr[:])

print("----------------------")

# Negative slicing
print(arr[-3:-1])

print("----------------------")

# Step
print(arr[::2])

print("----------------------")

# Reverse
print(arr[::-1])

print("----------------------")

# ==========================================
# 2D Array
# ==========================================

arr2 = np.array([
    [1,2,3,4],
    [5,6,7,8]
])

print(arr2)

print("----------------------")

# First row
print(arr2[0,1:3])

# Second row
print(arr2[1,0:2])

# Second column
print(arr2[:,1])

# Column slice
print(arr2[:,1:3])
````

## NumPy Data Types

### What is a Data Type?

- A data type defines the type of values stored in an array.
- Every NumPy array has a data type.

### Check Data Type

```python
print(arr.dtype)
```

### Common Data Type Codes

| Code | Meaning |
|------|---------|
| `i` | Integer |
| `f` | Float |
| `b` | Boolean |
| `U` | Unicode String |
| `S` | Byte String |
| `O` | Object |

### Set Data Type

```python
arr = np.array([1,2,3], dtype="f")
```

### Convert Data Type

```python
new_arr = arr.astype(float)
```

### Common Conversions

- Integer → Float
- Float → Integer
- Integer → Boolean
- Boolean → Integer

### Best Practice

- Use `arr.dtype` to check the data type.
- Use `astype()` to convert data types safely.
- Make sure the values can actually be converted.

````python
import numpy as np

# ==========================================
# Integer Array
# ==========================================

arr1 = np.array([10, 20, 30])

print(arr1)
print("Data Type:", arr1.dtype)

print("----------------------")

# ==========================================
# Float Array
# ==========================================

arr2 = np.array([10.5, 20.2, 30.8])

print(arr2)
print("Data Type:", arr2.dtype)

print("----------------------")

# ==========================================
# String Array
# ==========================================

arr3 = np.array(["Apple", "Mango", "Banana"])

print(arr3)
print("Data Type:", arr3.dtype)

print("----------------------")

# ==========================================
# Set Data Type
# ==========================================

arr4 = np.array([1, 2, 3], dtype="f")

print(arr4)
print("Data Type:", arr4.dtype)

print("----------------------")

# ==========================================
# Convert Integer to Float
# ==========================================

arr5 = np.array([10, 20, 30])

float_arr = arr5.astype(float)

print(float_arr)
print(float_arr.dtype)

print("----------------------")

# ==========================================
# Convert Float to Integer
# ==========================================

arr6 = np.array([1.5, 2.8, 3.9])

int_arr = arr6.astype(int)

print(int_arr)
print(int_arr.dtype)

print("----------------------")

# ==========================================
# Convert Integer to Boolean
# ==========================================

arr7 = np.array([1, 0, 5])

bool_arr = arr7.astype(bool)

print(bool_arr)
print(bool_arr.dtype)
````

## NumPy Array Reshape

### What is Reshape?

- Changes the shape of an array.
- The data remains the same.
- Only the arrangement changes.

### Syntax

```python
arr.reshape(rows, columns)
```

### 1D to 2D

```python
arr.reshape(2,3)
```

### 1D to 3D

```python
arr.reshape(2,2,2)
```

### Rule

- Total number of elements must remain the same.

Example:

- `2 × 3 = 6` ✅
- `3 × 2 = 6` ✅
- `4 × 2 = 8` ❌ (if array has only 6 elements)

### Unknown Dimension

```python
arr.reshape(2,-1)
```

- `-1` lets NumPy calculate the missing dimension automatically.

### Best Practice

- Check the array shape before reshaping.
- Ensure the total number of elements does not change.

````python
import numpy as np

# Original 1D Array
arr = np.array([1, 2, 3, 4, 5, 6])

print("Original Shape:", arr.shape)

# 1D -> 2D
arr2 = arr.reshape(2, 3)

print(arr2)
print("Shape:", arr2.shape)

print("------------------")

# 1D -> 3D
arr3 = np.array([1, 2, 3, 4, 5, 6, 7, 8])

new_arr = arr3.reshape(2, 2, 2)

print(new_arr)
print("Shape:", new_arr.shape)

print("------------------")

# Using -1
auto_shape = arr.reshape(2, -1)

print(auto_shape)
print("Shape:", auto_shape.shape)
````

## NumPy Array Iterating

### What is Iterating?

- Iterating means accessing each element of an array one by one.

### Iterate 1D Array

```python
for x in arr:
    print(x)
```

### Iterate 2D Array

```python
for row in arr:
    for value in row:
        print(value)
```

### Iterate 3D Array

- Use nested loops for each dimension.

### Using `np.nditer()`

```python
for x in np.nditer(arr):
    print(x)
```

- Works for arrays of any dimension.

### Using `np.ndenumerate()`

```python
for index, value in np.ndenumerate(arr):
    print(index, value)
```

- Returns both index and value.

### Best Practice

- Use normal `for` loop for simple 1D arrays.
- Use `np.nditer()` for arrays with multiple dimensions.
- Use `np.ndenumerate()` when you also need indexes.

````python
import numpy as np

# ==========================================
# 1D Array
# ==========================================

arr1 = np.array([10, 20, 30])

print("1D Array:")
for x in arr1:
    print(x)

print("------------------")

# ==========================================
# 2D Array
# ==========================================

arr2 = np.array([
    [1, 2, 3],
    [4, 5, 6]
])

print("2D Array:")
for row in arr2:
    for value in row:
        print(value)

print("------------------")

# ==========================================
# nditer()
# ==========================================

print("Using nditer:")
for x in np.nditer(arr2):
    print(x)

print("------------------")

# ==========================================
# ndenumerate()
# ==========================================

print("Using ndenumerate:")
for index, value in np.ndenumerate(arr2):
    print(index, value)
````

## NumPy Array Join

### What is Join?

- Join combines two or more arrays into one array.

### concatenate()

```python
np.concatenate((arr1, arr2))
```

### axis

- `axis=0` → Join rows
- `axis=1` → Join columns

### stack()

```python
np.stack((arr1, arr2))
```

- Creates a new dimension.

### hstack()

```python
np.hstack((arr1, arr2))
```

- Horizontal join.

### vstack()

```python
np.vstack((arr1, arr2))
```

- Vertical join.

### dstack()

```python
np.dstack((arr1, arr2))
```

- Joins along the third dimension.

### Best Practice

- Use `concatenate()` for normal joining.
- Use `stack()` when you need an extra dimension.

````python
import numpy as np

# ==========================================
# concatenate()
# ==========================================

arr1 = np.array([1, 2, 3])
arr2 = np.array([4, 5, 6])

print("concatenate():")
print(np.concatenate((arr1, arr2)))

print("------------------")

# ==========================================
# axis example
# ==========================================

a = np.array([[1, 2], [3, 4]])
b = np.array([[5, 6], [7, 8]])

print("axis=0")
print(np.concatenate((a, b), axis=0))

print()

print("axis=1")
print(np.concatenate((a, b), axis=1))

print("------------------")

# ==========================================
# stack()
# ==========================================

print("stack()")
print(np.stack((arr1, arr2)))

print("------------------")

# ==========================================
# hstack()
# ==========================================

print("hstack()")
print(np.hstack((arr1, arr2)))

print("------------------")

# ==========================================
# vstack()
# ==========================================

print("vstack()")
print(np.vstack((arr1, arr2)))

print("------------------")

# ==========================================
# dstack()
# ==========================================

print("dstack()")
print(np.dstack((arr1, arr2)))
````

## NumPy Array Split

### What is Split?

- Split divides one array into multiple smaller arrays.

### array_split()

```python
np.array_split(arr, parts)
```

### Example

```python
parts = np.array_split(arr, 3)
```

### Access Individual Part

```python
parts[0]
```

### 2D Split

- By default, splits rows.

### hsplit()

```python
np.hsplit(arr, 2)
```

- Splits columns.

### vsplit()

```python
np.vsplit(arr, 2)
```

- Splits rows.

### Difference

- `split()` → Equal split only.
- `array_split()` → Equal or unequal split.

### Best Practice

- Prefer `array_split()` because it is more flexible.

````python
import numpy as np

# ==========================================
# Split 1D Array
# ==========================================

arr = np.array([10, 20, 30, 40, 50, 60])

parts = np.array_split(arr, 3)

print("Split into 3 parts:")
print(parts)

print("------------------")

# Access first part
print("First Part:")
print(parts[0])

print("------------------")

# ==========================================
# Split 2D Array
# ==========================================

arr2 = np.array([
    [1, 2],
    [3, 4],
    [5, 6],
    [7, 8]
])

print("2D Split:")
print(np.array_split(arr2, 2))

print("------------------")

# ==========================================
# Horizontal Split
# ==========================================

arr3 = np.array([
    [1, 2, 3, 4],
    [5, 6, 7, 8]
])

print("Horizontal Split:")
print(np.hsplit(arr3, 2))

print("------------------")

# ==========================================
# Vertical Split
# ==========================================

print("Vertical Split:")
print(np.vsplit(arr3, 2))
````

## NumPy Array Search

### What is Search?

- Search is used to find the index of values in an array.

### `np.where()`

```python
np.where(condition)
```

- Returns indexes where the condition is True.

### Search a Value

```python
np.where(arr == 20)
```

### Search Using Condition

```python
np.where(arr > 25)
```

### Search Even Numbers

```python
np.where(arr % 2 == 0)
```

### Search Odd Numbers

```python
np.where(arr % 2 == 1)
```

- Remember that `np.where()` returns indexes, not values.
- It works with values as well as conditions.

## NumPy Filter Array

### What is Filtering?

- Filtering selects values based on a condition.
- It returns only the values that satisfy the condition.

### Boolean Mask

- A Boolean mask contains `True` and `False`.
- `True` → Select the value.
- `False` → Ignore the value.

### Filter Using Condition

```python
arr[arr > 30]
```

### Filter Even Numbers

```python
arr[arr % 2 == 0]
```

### Filter Odd Numbers

```python
arr[arr % 2 == 1]
```

### Custom Boolean Mask

```python
mask = [True, False, True, False]

arr[mask]
```

### `where()` vs Filtering

- `np.where()` → Returns indexes.
- Filtering → Returns values.

### Best Practice

- Use filtering when you need the actual data.
- Use `np.where()` when you need the indexes.

````python
import numpy as np

# ==========================================
# Original Array
# ==========================================

arr = np.array([10, 20, 30, 40, 50])

# Filter values greater than 30
print("Values > 30:")
print(arr[arr > 30])

print("------------------")

# ==========================================
# Even Numbers
# ==========================================

arr2 = np.array([1, 2, 3, 4, 5, 6])

print("Even Numbers:")
print(arr2[arr2 % 2 == 0])

print("------------------")

# ==========================================
# Odd Numbers
# ==========================================

print("Odd Numbers:")
print(arr2[arr2 % 2 == 1])

print("------------------")

# ==========================================
# Boolean Mask
# ==========================================

arr3 = np.array([10, 20, 30, 40])

mask = [True, False, True, False]

print("Filtered with Boolean Mask:")
print(arr3[mask])
````
