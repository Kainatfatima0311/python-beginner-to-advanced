###Arguments
**Samajhne wali baatein:**

Argument woh value hoti hai jo function call karte waqt bhejte hain.

Parameter woh variable hota hai jo function define karte waqt brackets mein likhte hain.

Function ko jitne arguments chahiye, utne hi dene hote hain.

Default value de do to argument optional ho jata hai.

Positional arguments mein order matter karta hai.

Keyword arguments mein key = value likhte hain, order matter nahi karta.

Positional argument hamesha keyword argument se pehle aata hai.

Function mein string, number, list, dictionary sab argument ke tor par bhej sakte hain.

Function list, tuple, dictionary waghera bhi return kar sakta hai.

"*" ka matlab: argument sirf keyword se dena hoga.

````python
# Python Function Arguments

# parameter: fname
# argument: "Ali"
def full_name(fname, lname):
    print(fname, lname)


full_name("Ali", "Khan")
# full_name("Ali")  # error: 2 arguments chahiye


# default parameter
def greet(name="friend"):
    print("Hello", name)


greet("Ahmed")
greet()


# positional arguments
def pet_info(animal, name):
    print("I have a", animal)
    print("Its name is", name)


pet_info("dog", "Buddy")


# keyword arguments
pet_info(name="Kitty", animal="cat")


# mix positional + keyword
def student(name, age, city):
    print(name, age, city)


student("Ali", age=20, city="Lahore")


# list as argument
def show_fruits(fruits):
    for fruit in fruits:
        print(fruit)


my_fruits = ["apple", "banana", "cherry"]
show_fruits(my_fruits)


# dictionary as argument
def show_person(person):
    print("Name:", person["name"])
    print("Age:", person["age"])


my_person = {"name": "Emil", "age": 25}
show_person(my_person)


# return value
def add(x, y):
    return x + y


result = add(5, 3)
print(result)


# return list
def get_fruits():
    return ["apple", "banana", "cherry"]


fruits = get_fruits()
print(fruits[0])


# return tuple
def get_numbers():
    return (10, 20)


x, y = get_numbers()
print(x)
print(y)


# positional-only argument
def say_name(name, /):
    print("Hello", name)


say_name("Ali")
# say_name(name="Ali")  # error


# keyword-only argument
def say_city(*, city):
    print("City:", city)


say_city(city="Karachi")
# say_city("Karachi")  # error


# positional-only + keyword-only
def calculate(a, b, /, *, c, d):
    return a + b + c + d


answer = calculate(5, 10, c=15, d=20)
print(answer)
````

##Python *args and **kwargs

*args unknown number of positional arguments ke liye hota hai.

*args function ke andar tuple ban jata hai.

**kwargs unknown number of keyword arguments ke liye hota hai.

**kwargs function ke andar dictionary ban jata hai.

Regular parameter hamesha *args aur **kwargs se pehle aata hai.

Dono sath use karne ka order hota hai: normal, *args, **kwargs.

Function define karte waqt * aur ** arguments ko collect karte hain.

Function call karte waqt * list/tuple ko unpack karta hai.

Function call karte waqt ** dictionary ko unpack karta hai.

````python
# Python *args and **kwargs

# *args = multiple positional arguments
def show_kids(*kids):
    print(type(kids))
    print(kids)
    print(kids[2])


show_kids("Emil", "Tobias", "Linus")


# regular parameter + *args
def greet_all(greeting, *names):
    for name in names:
        print(greeting, name)


greet_all("Hello", "Ali", "Ahmed", "Sara")


# practical use of *args
def add_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total


print(add_numbers(1, 2, 3))
print(add_numbers(10, 20, 30, 40))


# maximum value with *args
def find_max(*numbers):
    if len(numbers) == 0:
        return None

    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num

    return max_num


print(find_max(3, 7, 2, 9, 1))


# **kwargs = multiple keyword arguments
def show_person(**person):
    print(type(person))
    print(person)
    print(person["name"])
    print(person["age"])


show_person(name="Tobias", age=30, city="Bergen")


# regular parameter + **kwargs
def user_details(username, **details):
    print("Username:", username)

    for key, value in details.items():
        print(key, value)


user_details("emil123", age=25, city="Oslo", hobby="coding")


# normal + *args + **kwargs
def full_info(title, *args, **kwargs):
    print("Title:", title)
    print("Positional:", args)
    print("Keyword:", kwargs)


full_info("User Info", "Emil", "Tobias", age=25, city="Oslo")


# unpack list with *
def add_three(a, b, c):
    return a + b + c


numbers = [1, 2, 3]
print(add_three(*numbers))


# unpack dictionary with **
def full_name(fname, lname):
    print("Hello", fname, lname)


person = {"fname": "Emil", "lname": "Refsnes"}
full_name(**person)
````

## Python Scope

### What is Scope?

- **Scope** batata hai ke koi variable program ke kis part mein access (use) ho sakta hai.
- Agar variable function ke andar bana hai to uska scope alag hoga.
- Agar function ke bahar bana hai to uska scope alag hoga.
- Python variable ko dhoondhne ke liye **LEGB Rule** follow karta hai.

**LEGB Rule**

- **L = Local**
- **E = Enclosing**
- **G = Global**
- **B = Built-in**

Python isi order mein variable ko search karta hai.

---

### Local Scope

- Function ke andar banaya gaya variable **Local Variable** kehlata hai.
- Is variable ko sirf usi function ke andar access kiya ja sakta hai.
- Function ke bahar use karne par **NameError** aata hai.

---

### Function Inside Function

- Python mein ek function ke andar doosra function bhi bana sakte hain.
- Inner function outer function ke variables ko access kar sakta hai.
- Is scope ko **Enclosing Scope** kehte hain.

---

### Local Scope Example

- Agar inner function ke andar same naam ka variable bana diya jaye to Python pehle us local variable ko use karega.
- Python hamesha sabse pehle Local Scope check karta hai.

---

### Global Scope

- Function ke bahar banaya gaya variable **Global Variable** hota hai.
- Is variable ko program ke kisi bhi function ke andar access kiya ja sakta hai.
- Jab tak usi naam ka local variable na ban jaye, global variable hi use hota hai.

---

### Global Keyword

- Agar function ke andar global variable ki value change karni ho to **global** keyword use karte hain.
- Agar global keyword use na karein to Python naya local variable samajhta hai aur error aa sakta hai.

---

### Nonlocal Keyword

- **nonlocal** nested functions mein use hota hai.
- Iski madad se outer function ke variable ki value change ki ja sakti hai.
- Ye sirf enclosing scope ke variables ke liye use hota hai.

---

### Built-in Scope

- Python mein kuch functions aur names pehle se available hote hain.
- Inhein **Built-in Scope** kehte hain.

Examples:

- print()
- len()
- list()
- sum()
- max()
- min()
- type()

---

### Built-in Names Ko Overwrite Na Karein

- Kabhi bhi variables ke naam built-in functions ke naam par na rakhein.

Examples:

- list
- str
- int
- type
- sum

- Agar aisa karenge to original built-in function overwrite ho jayega aur baad mein errors aa sakte hain.

Example Error:

`TypeError: 'list' object is not callable`

---

### LEGB Rule Summary

Python variable ko is order mein search karta hai:

1. Local
2. Enclosing
3. Global
4. Built-in

Jahan variable mil jaye Python wahi search stop kar deta hai.

---

### Best Practice

- Global variables kam use karein.
- Functions ke andar data pass karne ke liye parameters aur return values use karna better practice hai.
- Isse code clean aur maintainable rehta hai.

````python
# ==========================
# GLOBAL VARIABLE
# ==========================

message = "Hello Python"   # Ye Global Variable hai
counter = 0                # Is variable ko poore program mein access kar sakte hain


# ==========================
# LOCAL + ENCLOSING + NONLOCAL SCOPE
# ==========================

def outer():

    topic = "Python Scope"      # Ye Enclosing Variable hai (outer function ka)

    def inner():

        nonlocal topic
        # nonlocal ka matlab:
        # Outer function ke variable ko modify karna

        topic = "Python Scope Updated"

        local_var = "I am Local"
        # Ye sirf inner function ke andar available hai

        print(local_var)
        print(topic)

    inner()     # Inner function call

    # nonlocal ki wajah se topic ki value update ho chuki hai
    print(topic)


outer()


# ==========================
# GLOBAL KEYWORD
# ==========================

def increase_counter():

    global counter
    # global keyword batata hai ke hum global variable use kar rahe hain

    counter += 1
    # Global counter ki value increase ho gayi


increase_counter()

print(counter)


# ==========================
# ACCESSING GLOBAL VARIABLE
# ==========================

def show_message():

    # message function ke bahar bana hai
    # Lekin function ke andar bhi access ho raha hai
    print(message)


show_message()


# ==========================
# BUILT-IN FUNCTIONS
# ==========================

numbers = [10, 20, 30]

# len() list ki total length batata hai
print(len(numbers))

# sum() list ka total nikalta hai
print(sum(numbers))

# max() sabse badi value return karta hai
print(max(numbers))

# type() object ka data type batata hai
print(type(numbers))
````

## Python Decorators

### What are Decorators?

- Decorator ek special function hota hai jo kisi doosre function ki functionality ko **modify ya extend** karta hai.
- Original function ka code change kiye bina usmein extra functionality add ki ja sakti hai.
- Python mein functions bhi objects hote hain, isi wajah se decorators possible hain.

---

### Functions are Objects

- Python mein function ko variable mein store kar sakte hain.
- Function ko argument ke taur par kisi doosre function ko pass kar sakte hain.
- Function ko return bhi kar sakte hain.
- Ye concept decorators ki foundation hai.

---

### Create a Decorator

Decorator banane ke liye normally 3 cheezen hoti hain:

- **Decorator Function** → Jo original function ko receive karta hai.
- **Wrapper Function** → Original function ke around extra code execute karta hai.
- **Return Wrapper** → Wrapper function ko return kiya jata hai.

Flow:

Decorator → Wrapper → Original Function

---

### Why Use Decorators?

Decorators ka use repeated code ko avoid karne ke liye hota hai.

Common uses:

- Logging
- Authentication
- Authorization
- Input Validation
- Performance Timing
- Error Handling

---

### @ Decorator Syntax

Decorator ko apply karne ke do tareeqe hain.

Without @

- Function ko manually decorator ke through wrap karte hain.

With @

- `@decorator_name` shortcut syntax hai.
- Ye automatically function ko decorator ke through wrap kar deta hai.
- Isi syntax ko zyada use kiya jata hai.

---

### Multiple Decorators

- Ek function par ek se zyada decorators bhi laga sakte hain.
- Python decorators ko **bottom se top** apply karta hai.

Example:

@decorator1
@decorator2

Function call hone par pehle **decorator2** execute hoga, phir **decorator1**.

---

### Benefits of Decorators

- Code Reusability
- Clean Code
- Less Repetition
- Easy Maintenance
- Better Readability

---

### Best Practice

- Jab same functionality multiple functions mein repeat ho rahi ho tab decorator use karein.
- Sirf zarurat ke waqt decorators use karein.
- Wrapper function ko simple aur readable rakhein.

````python
# =====================================
# 1. Functions are Objects
# =====================================

def greet():
    print("Hello Python!")

# Function ko variable mein store kiya
say_hello = greet

# Dono same function ko call karenge
say_hello()


# =====================================
# 2. Function as Argument
# =====================================

def execute(func):
    # func ek function receive kar raha hai
    print("Executing Function...")
    func()   # Received function ko call kiya

# greet function ko argument ke taur par pass kiya
execute(greet)


# =====================================
# 3. Simple Decorator
# =====================================

def my_decorator(func):
    # func original function hai

    def wrapper():
        # Original function se pehle chalega
        print("Before Function")

        # Original function execute hoga
        func()

        # Original function ke baad chalega
        print("After Function")

    # Wrapper return karna zaroori hai
    return wrapper


# =====================================
# 4. Without @ Syntax
# =====================================

def welcome():
    print("Welcome!")

# Manually decorator apply kiya
decorated_function = my_decorator(welcome)

decorated_function()


# =====================================
# 5. Using @ Syntax
# =====================================

@my_decorator
def learning():
    print("Learning Python Decorators")

# Python automatically:
# learning = my_decorator(learning)

learning()


# =====================================
# 6. Multiple Decorators
# =====================================

def stars(func):

    def wrapper():
        print("*" * 20)

        func()

        print("*" * 20)

    return wrapper


def smile(func):

    def wrapper():
        print("😊")

        func()

        print("😊")

    return wrapper


# Python pehle smile apply karega
# Phir stars apply karega

@stars
@smile
def message():
    print("Decorators are Awesome!")

message()
````

## Python Lambda

### What is Lambda?

- **Lambda** ek anonymous (naam ke bina) function hota hai.
- Iska use chhote aur simple functions banane ke liye hota hai.
- Agar function sirf ek hi expression perform karta ho to lambda use karna convenient hota hai.

Syntax:

```python
lambda arguments : expression
```

- `lambda` → keyword
- `arguments` → function ke parameters
- `expression` → jo value return hogi

> Lambda function mein `return` keyword nahi likhte.
> Expression ki value automatically return ho jati hai.

---

### Why Use Lambda?

- Chhota aur clean code likhne ke liye.
- Ek hi baar use hone wale functions ke liye.
- Higher-order functions (jaise `map()`, `filter()`, `sorted()`) ke sath bohat use hota hai.

---

### Lambda Function with One Argument

- Lambda ek argument bhi receive kar sakta hai.
- Expression execute karke result return karta hai.

Example idea:

```
lambda x : x + 10
```

---

### Lambda Function with Multiple Arguments

- Lambda function multiple arguments bhi le sakta hai.
- Arguments comma `,` se separate hote hain.

Example idea:

```
lambda x, y : x * y
```

---

### Why Use Lambda Functions?

W3Schools ke mutabiq lambda ka use tab hota hai jab:

- Function sirf aik dafa use hona ho.
- Chhota function banana ho.
- Kisi function ke andar temporary function banana ho.

---

### Lambda Inside Another Function

- Ek function lambda return bhi kar sakta hai.
- Is tarah dynamic functions bana sakte hain.
- Ye technique reusable code likhne mein help karti hai.

---

### Difference Between def and lambda

#### def Function

- Proper named function hota hai.
- Multiple statements likh sakte hain.
- Complex logic ke liye use hota hai.

#### Lambda Function

- Anonymous function hota hai.
- Sirf **one expression** likh sakte hain.
- Chhote tasks ke liye best hai.

---

### Limitations of Lambda

- Sirf **one expression** allow hota hai.
- Loops ya multiple statements directly nahi likh sakte.
- Large functions ke liye `def` use karna better hota hai.

---

### Best Practice

- Lambda sirf simple operations ke liye use karein.
- Agar function bada ho raha hai to `def` use karein.
- Readability ko hamesha priority dein.

````python
# =====================================
# Normal Function (def)
# =====================================

# Normal function ka naam hota hai
def square(num):
    return num * num

print(square(5))


# =====================================
# Simple Lambda Function
# =====================================

# Lambda function ka koi naam nahi hota
square_lambda = lambda num: num * num

# Lambda automatically expression ki value return karta hai
print(square_lambda(5))


# =====================================
# Lambda with One Argument
# =====================================

add_ten = lambda x: x + 10

# 5 + 10
print(add_ten(5))


# =====================================
# Lambda with Two Arguments
# =====================================

multiply = lambda x, y: x * y

# 4 × 6
print(multiply(4, 6))


# =====================================
# Lambda with Three Arguments
# =====================================

calculate = lambda a, b, c: a + b + c

print(calculate(10, 20, 30))


# =====================================
# Lambda Inside a Function
# =====================================

def multiplier(n):
    # Lambda function return ho raha hai
    return lambda a: a * n

# n = 2
double = multiplier(2)

# n = 3
triple = multiplier(3)

print(double(8))   # 8 × 2
print(triple(8))   # 8 × 3


# =====================================
# Lambda with sorted()
# =====================================

students = [
    ("Ali", 85),
    ("Ahmed", 70),
    ("Sara", 95)
]

# Score ke basis par sorting ho rahi hai
# item[1] second value (marks) ko represent karta hai
sorted_students = sorted(students, key=lambda item: item[1])

print(sorted_students)


# =====================================
# Lambda with filter()
# =====================================

numbers = [10, 15, 20, 25, 30]

# Sirf even numbers filter honge
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))

print(even_numbers)


# =====================================
# Lambda with map()
# =====================================

numbers = [1, 2, 3, 4, 5]

# Har number ka square banega
squares = list(map(lambda x: x * x, numbers))

print(squares)
````

## Python Recursion

### What is Recursion?

- **Recursion** ka matlab hai jab koi function **khud ko hi call kare**.
- Function apna kaam complete karne ke liye baar baar khud ko call karta rehta hai.
- Recursion tab tak chalti rehti hai jab tak **base case** complete na ho jaye.

---

### Base Case

- Har recursive function mein **Base Case** hona bohat zaroori hai.
- Base Case recursion ko stop karta hai.
- Agar Base Case na ho to function hamesha chalta rahega aur error aa jayega.

Error:

```
RecursionError: maximum recursion depth exceeded
```

---

### Recursive Case

- Base Case ke ilawa jo function dobara khud ko call karta hai usay **Recursive Case** kehte hain.
- Har call mein problem thori chhoti hoti jani chahiye.

---

### How Recursion Works?

Flow:

1. Function call hota hai.
2. Function khud ko dobara call karta hai.
3. Ye process repeat hoti rehti hai.
4. Jab Base Case milta hai to recursion ruk jati hai.
5. Phir saare function calls reverse order mein complete hote hain.

---

### Why Use Recursion?

Recursion ka use:

- Tree Traversal
- Folder Searching
- Factorial
- Fibonacci Series
- Backtracking Algorithms

mein bohat hota hai.

---

### Advantages

- Code chhota hota hai.
- Complex problems ko simple bana deta hai.
- Recursive problems ko solve karna easy hota hai.

---

### Disadvantages

- Zyada memory use karta hai.
- Har function call stack mein store hoti hai.
- Kabhi kabhi loop se slow hota hai.
- Base Case na ho to program crash kar sakta hai.

---

### Best Practice

- Hamesha Base Case likhein.
- Recursive call ko Base Case ki taraf move karna chahiye.
- Simple problems ke liye loop use karna better hota hai.
- Recursion tab use karein jab problem naturally recursive ho.

````python
# ==============================
# Simple Recursion Example
# ==============================

def my_function(num):

    # Base Case
    # Jab num 0 ho jaye to recursion stop ho jayegi
    if num == 0:
        print("Recursion End")
        return

    # Current value print hogi
    print(num)

    # Function khud ko dobara call kar raha hai
    my_function(num - 1)


# Function call
my_function(5)


print("----------------------")


````

````python
# ==============================
# Factorial Using Recursion
# ==============================

def factorial(num):

    # Base Case
    if num == 1:
        return 1

    # Recursive Case
    # Example:
    # 5 × factorial(4)
    return num * factorial(num - 1)


print("Factorial =", factorial(5))


print("----------------------")


# ==============================
# Compare Loop vs Recursion
# ==============================

print("Using Loop")

for num in range(5, 0, -1):
    print(num)


print("----------------------")

print("Using Recursion")

my_function(5)
````

## Python Generators

### What is a Generator?

- **Generator** ek special function hota hai jo values **ek ek karke** return karta hai.
- Normal function `return` use karta hai, jabke Generator `yield` keyword use karta hai.
- Generator sari values ek sath memory mein store nahi karta.
- Jab value ki zarurat hoti hai tabhi next value generate karta hai.

---

### Generator vs Normal Function

#### Normal Function

- `return` keyword use karta hai.
- Ek baar value return karta hai.
- Function khatam ho jata hai.
- Dobara value chahiye to function phir call karna parta hai.

#### Generator Function

- `yield` keyword use karta hai.
- Ek ek value return karta hai.
- Function apni previous position yaad rakhta hai.
- Jahan se stop hua tha wahi se next value generate karta hai.

---

### Creating a Generator

- Generator function bhi `def` se hi banta hai.
- Farq sirf itna hai ke `return` ki jagah `yield` use hota hai.
- Jis function mein `yield` ho wo Generator Function ban jata hai.

---

### Yield Keyword

- `yield` value return bhi karta hai aur function ko pause bhi kar deta hai.
- Jab next value mangi jati hai to function wahi se continue hota hai jahan pause hua tha.
- Is wajah se generators memory efficient hote hain.

---

### Iterate Through Generator

Generator ki values ko access karne ke liye:

- `for` loop use kar sakte hain.
- Ya `next()` function use kar sakte hain.

---

### Why Use Generators?

Generators tab use hote hain jab:

- Bohat zyada data ho.
- Memory bachani ho.
- Data ko ek ek karke process karna ho.

Examples:

- Large Files
- Data Streaming
- Infinite Sequences
- Reading Logs

---

### Generator Expression

- Jaise List Comprehension hoti hai waise hi Generator Expression bhi hoti hai.
- Difference ye hai ke List memory mein sari values store karti hai.
- Generator sirf zarurat par next value generate karta hai.

List:

```python
[x for x in range(5)]
```

Generator:

```python
(x for x in range(5))
```

---

### Advantages

- Memory efficient.
- Large data handle kar sakta hai.
- Fast execution.
- Lazy Evaluation use karta hai (value tab banti hai jab zarurat ho).

---

### Disadvantages

- Values sirf ek baar iterate hoti hain.
- Random indexing possible nahi hoti.
- Agar dobara use karna ho to generator phir se banana parta hai.

---

### Best Practice

- Large datasets ke liye generators use karein.
- Chhoti lists ke liye normal list hi better hoti hai.
- Jab memory important ho tab generator use karein.

````python
# ==========================
# Normal Function
# ==========================

def my_numbers():

    # return sirf ek hi baar value deta hai
    return 1


print(my_numbers())

print("----------------")


# ==========================
# Generator Function
# ==========================

def my_generator():

    # Pehli value
    yield 1

    # Function yahan pause hoga

    # Dusri value
    yield 2

    # Phir pause hoga

    # Teesri value
    yield 3


# Generator object ban gaya
numbers = my_generator()

# next() se ek ek value milegi
print(next(numbers))   # 1

print(next(numbers))   # 2

print(next(numbers))   # 3


print("----------------")


# ==========================
# Using for Loop
# ==========================

def fruits():

    yield "Apple"

    yield "Mango"

    yield "Banana"


# for loop automatically next() call karta hai
for item in fruits():

    print(item)


print("----------------")


# ==========================
# Generator Expression
# ==========================

# Generator Expression
numbers = (num for num in range(1, 6))

for num in numbers:

    print(num)


print("----------------")


# ==========================
# List vs Generator
# ==========================

# List saari values memory mein store karti hai
my_list = [num for num in range(1, 6)]

print(my_list)


# Generator ek ek value banata hai
my_generator = (num for num in range(1, 6))

print(my_generator)

# Generator ki values dekhne ke liye loop use karein
for num in my_generator:

    print(num)
````
