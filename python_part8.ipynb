##Python While Loops

###Python Loops
Python has two primitive loop commands:

while loops
for loops

```python
i = 1
while i < 6:
  print(i)
  i += 1
```

```python
i = 1
while i < 6:
  print(i)
  if i == 3:
    break #break loop ko foran stop kar deta hai.
  i += 1
```

```python
i = 0

while i < 5:
    i += 1

    if i == 3:
        continue #current miss kr k next pr chala jata hai

    print(i)
```

```python
i = 1

while i <= 3:
    print(i)
    i += 1
else:
    print("Loop finished")
```

##Python For Loops
A for loop is used for iterating over a sequence (that is either a list, a tuple, a dictionary, a set, or a string).

This is less like the for keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages.

With the for loop we can execute a set of statements, once for each item in a list, tuple, set etc.

```python
#for loop tab use hota hai jab tum kisi
#sequence ke har item par kaam karna chahte ho.
fruits = ["apple", "banana", "cherry"]
for x in fruits:
  print(x)
```

```python
for letter in "banana":
    print(letter)
```

```python
for x in range(2, 6):
    print(x)
```

```python
#start,stop,step
for x in range(2, 10, 2):
    print(x)
```

```python
fruits = ["apple", "banana", "cherry"]

for x in fruits:
    if x == "banana":
        break
    print(x)
```

```python
fruits = ["apple", "banana", "cherry"]

for x in fruits:
    if x == "banana":
        continue
    print(x)
```

```python
for x in range(3):
    print(x)
else:
    print("Finished")
```

```python
colors = ["red", "green"]
fruits = ["apple", "banana"]

for color in colors:
    for fruit in fruits:
        print(color, fruit)
```

##Python Functions
A function is a block of code which only runs when it is called.

A function can return data as a result.

A function helps avoiding code repetition.

```python
def add_numbers():
    return 5 + 3

result = add_numbers()
print(result)
```

```python
# Python Functions

# simple function
def say_hi():
    print("Hi, I am learning Python")


say_hi()


# function with parameter
def greet(name):
    print("Hello", name)


greet("Ali")
greet("Ahmed")


# function with return
def add_numbers(a, b):
    return a + b


result = add_numbers(5, 3)
print(result)


# return value used directly
def multiply(a, b):
    return a * b


print(multiply(4, 5))


# empty function
def empty_function():
    pass
```

###Arguments
**Samajhne wali baatein:**

Argument woh value hoti hai jo function call karte waqt bhejte hain.

Parameter woh variable hota hai jo function define karte waqt brackets mein likhte hain.

Function ko jitne arguments chahiye, utne hi dene hote hain.

Default value de do to argument optional ho jata hai.

Positional arguments mein order matter karta hai.

Keyword arguments mein key = value likhte hain, order matter nahi karta.

Positional argument hamesha keyword argument se pehle aata hai.

Function mein string, number, list, dictionary sab argument ke tor par bhej sakte hain.

return function se result wapas deta hai.

Function list, tuple, dictionary waghera bhi return kar sakta hai.

/ ka matlab: argument sirf positional hoga.

"*" ka matlab: argument sirf keyword se dena hoga.

```python
# Python Function Arguments

# parameter: fname
# argument: "Ali"
def full_name(fname, lname):
    print(fname, lname)


full_name("Ali", "Khan")
# full_name("Ali")  # error: 2 arguments chahiye


# default parameter
def greet(name="friend"):
    print("Hello", name)


greet("Ahmed")
greet()


# positional arguments
def pet_info(animal, name):
    print("I have a", animal)
    print("Its name is", name)


pet_info("dog", "Buddy")


# keyword arguments
pet_info(name="Kitty", animal="cat")


# mix positional + keyword
def student(name, age, city):
    print(name, age, city)


student("Ali", age=20, city="Lahore")


# list as argument
def show_fruits(fruits):
    for fruit in fruits:
        print(fruit)


my_fruits = ["apple", "banana", "cherry"]
show_fruits(my_fruits)


# dictionary as argument
def show_person(person):
    print("Name:", person["name"])
    print("Age:", person["age"])


my_person = {"name": "Emil", "age": 25}
show_person(my_person)


# return value
def add(x, y):
    return x + y


result = add(5, 3)
print(result)


# return list
def get_fruits():
    return ["apple", "banana", "cherry"]


fruits = get_fruits()
print(fruits[0])


# return tuple
def get_numbers():
    return (10, 20)


x, y = get_numbers()
print(x)
print(y)


# positional-only argument
def say_name(name, /):
    print("Hello", name)


say_name("Ali")
# say_name(name="Ali")  # error


# keyword-only argument
def say_city(*, city):
    print("City:", city)


say_city(city="Karachi")
# say_city("Karachi")  # error


# positional-only + keyword-only
def calculate(a, b, /, *, c, d):
    return a + b + c + d


answer = calculate(5, 10, c=15, d=20)
print(answer)
```

##Python *args and **kwargs

*args unknown number of positional arguments ke liye hota hai.

*args function ke andar tuple ban jata hai.

**kwargs unknown number of keyword arguments ke liye hota hai.

**kwargs function ke andar dictionary ban jata hai.

Regular parameter hamesha *args aur **kwargs se pehle aata hai.

Dono sath use karne ka order hota hai: normal, *args, **kwargs.

Function define karte waqt * aur ** arguments ko collect karte hain.

Function call karte waqt * list/tuple ko unpack karta hai.

Function call karte waqt ** dictionary ko unpack karta hai.

```python
# Python *args and **kwargs

# *args = multiple positional arguments
def show_kids(*kids):
    print(type(kids))
    print(kids)
    print(kids[2])


show_kids("Emil", "Tobias", "Linus")


# regular parameter + *args
def greet_all(greeting, *names):
    for name in names:
        print(greeting, name)


greet_all("Hello", "Ali", "Ahmed", "Sara")


# practical use of *args
def add_numbers(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total


print(add_numbers(1, 2, 3))
print(add_numbers(10, 20, 30, 40))


# maximum value with *args
def find_max(*numbers):
    if len(numbers) == 0:
        return None

    max_num = numbers[0]
    for num in numbers:
        if num > max_num:
            max_num = num

    return max_num


print(find_max(3, 7, 2, 9, 1))


# **kwargs = multiple keyword arguments
def show_person(**person):
    print(type(person))
    print(person)
    print(person["name"])
    print(person["age"])


show_person(name="Tobias", age=30, city="Bergen")


# regular parameter + **kwargs
def user_details(username, **details):
    print("Username:", username)

    for key, value in details.items():
        print(key, value)


user_details("emil123", age=25, city="Oslo", hobby="coding")


# normal + *args + **kwargs
def full_info(title, *args, **kwargs):
    print("Title:", title)
    print("Positional:", args)
    print("Keyword:", kwargs)


full_info("User Info", "Emil", "Tobias", age=25, city="Oslo")


# unpack list with *
def add_three(a, b, c):
    return a + b + c


numbers = [1, 2, 3]
print(add_three(*numbers))


# unpack dictionary with **
def full_name(fname, lname):
    print("Hello", fname, lname)


person = {"fname": "Emil", "lname": "Refsnes"}
full_name(**person)
```
