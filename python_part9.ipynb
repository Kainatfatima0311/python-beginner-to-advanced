## Python Range

### What is range()?

- `range()` Python ka built-in function hai.
- Iska use numbers ki sequence generate karne ke liye hota hai.
- Ye zyada tar `for` loop ke sath use hota hai.
- `range()` khud list return nahi karta, balki **range object** return karta hai.

---

### Syntax

```python
range(start, stop, step)
```

Parameters:

- **start** → Number kahan se shuru hoga. (Default = 0)
- **stop** → Number kahan tak jayega. (Ye include nahi hota.)
- **step** → Kitne numbers skip karne hain. (Default = 1)

---

### Using range()

Agar sirf ek value dein:

```python
range(5)
```

Output:

```
0 1 2 3 4
```

- Counting 0 se start hoti hai.
- Stop value include nahi hoti.

---

### Start Parameter

Agar start aur stop dono dein:

```python
range(2, 6)
```

Output:

```
2 3 4 5
```

- Counting start value se shuru hoti hai.
- Stop value phir bhi include nahi hoti.

---

### Step Parameter

Step batata hai ke har next value kitni increase hogi.

Example:

```python
range(2, 10, 2)
```

Output:

```
2 4 6 8
```

Yahan har baar 2 ka jump lag raha hai.

---

### Negative Step

Negative step se reverse counting hoti hai.

Example:

```python
range(10, 0, -2)
```

Output:

```
10 8 6 4 2
```

---

### Convert range() into List

Agar range ki sari values ek sath dekhni hon to `list()` use karte hain.

Example:

```python
list(range(5))
```

Output:

```
[0, 1, 2, 3, 4]
```

---

### Common Uses of range()

- for loop chalane ke liye.
- Fixed number of iterations.
- Number sequences generate karne ke liye.
- Reverse counting.
- Tables aur patterns banane ke liye.

---

### Best Practice

- Yaad rakho **stop value kabhi include nahi hoti**.
- Agar values print karni hon to `for` loop use karo.
- Agar list chahiye ho to `list(range())` use karo.

````python
# ==========================
# range() with One Argument
# ==========================

print("range(5)")

# 0 se start hoga
# 5 include nahi hoga
for num in range(5):
    print(num)


print("----------------")


# ==========================
# range() with Start & Stop
# ==========================

print("range(2, 6)")

# 2 se start
# 6 include nahi hoga
for num in range(2, 6):
    print(num)


print("----------------")


# ==========================
# range() with Step
# ==========================

print("range(2, 10, 2)")

# Har baar 2 ka jump lagega
for num in range(2, 10, 2):
    print(num)


print("----------------")


# ==========================
# Negative Step
# ==========================

print("range(10, 0, -2)")

# Reverse counting
for num in range(10, 0, -2):
    print(num)


print("----------------")


# ==========================
# Convert range() into List
# ==========================

# list() se range object ki sari values ek sath mil jati hain
numbers = list(range(5))

print(numbers)


print("----------------")


# ==========================
# Practical Example
# ==========================

print("Table of 2")

# 2 ka table print karna
for num in range(1, 11):

    print(f"2 x {num} = {2 * num}")
````

## Python Arrays

### What are Arrays?

- Python mein C++ ya Java ki tarah built-in Arrays nahi hoti.
- Python mein arrays ki jagah **Lists** zyada use hoti hain.
- Agar actual Array use karni ho to Python ka **array module** import karna parta hai.
- Array ek hi type ka multiple data store karti hai.

Example:

```
[10, 20, 30, 40]
```

Sab values integer hain.

---

### Create an Array

- Array banane ke liye pehle `array` module import karte hain.
- Phir `array()` function use hota hai.

Syntax:

```python
array.array(typecode, values)
```

---

### Type Code

Type code batata hai array kis type ka data store karegi.

Common Type Codes:

| Type Code | Data Type |
|-----------|-----------|
| i | Integer |
| f | Float |
| d | Double |

---

### Access Array Elements

- Index number se element access karte hain.
- Index **0** se start hota hai.

Example:

```
Index

0
1
2
3
```

---

### Modify Array Elements

- Kisi bhi index ki value change ki ja sakti hai.
- Sirf usi index par new value assign karni hoti hai.

---

### Array Length

- `len()` function array ke total elements batata hai.

---

### Loop Through an Array

- `for` loop use karke array ke tamam elements access kar sakte hain.

---

### Remove Elements

Array se element remove karne ke liye:

- `remove()`
- `pop()`

use kar sakte hain.

---

### Search an Array

- `index()` kisi value ka index batata hai.
- Agar value exist na kare to error aata hai.

---

### Best Practice

- General programming mein Lists zyada use hoti hain.
- Arrays tab use karein jab same data type ka data store karna ho.
- Arrays memory efficient hoti hain compared to lists.

````python
# Import array module
import array


# ==========================
# Create an Array
# ==========================

# 'i' means Integer Array
numbers = array.array('i', [10, 20, 30, 40])

print(numbers)


print("----------------")


# ==========================
# Access Elements
# ==========================

# Index starts from 0

print(numbers[0])     # First element
print(numbers[2])     # Third element


print("----------------")


# ==========================
# Modify Element
# ==========================

# Index 1 ki value change kar di

numbers[1] = 100

print(numbers)


print("----------------")


# ==========================
# Array Length
# ==========================

# Total elements count karega

print(len(numbers))


print("----------------")


# ==========================
# Loop Through Array
# ==========================

# Har element ek ek karke print hoga

for num in numbers:

    print(num)


print("----------------")


# ==========================
# Add New Element
# ==========================

# append() end mein value add karta hai

numbers.append(50)

print(numbers)


print("----------------")


# ==========================
# Remove Element
# ==========================

# remove() value ko delete karta hai

numbers.remove(30)

print(numbers)


print("----------------")


# ==========================
# Pop Element
# ==========================

# pop() last element remove karta hai

numbers.pop()

print(numbers)


print("----------------")


# ==========================
# Search Element
# ==========================

# index() value ka index batata hai

print(numbers.index(100))
````

## Python Iterators

### What is an Iterator?

- Iterator ek object hota hai jo data ko **ek ek karke** access karta hai.
- Iterator ki help se hum collection (List, Tuple, String, Dictionary, Set) ke elements ko one by one read kar sakte hain.
- Python mein har iterator `next()` function ki madad se next value return karta hai.

---

### Iterable vs Iterator

#### Iterable

- Jis object par loop chal sakta ho usay **Iterable** kehte hain.
- Examples:
  - List
  - Tuple
  - String
  - Dictionary
  - Set

#### Iterator

- Iterator wo object hota hai jo iterable ki values ek ek karke return karta hai.
- Iterator banane ke liye `iter()` function use hota hai.

---

### Create an Iterator

- `iter()` kisi iterable object ko iterator mein convert karta hai.
- Iterator banne ke baad `next()` se next value milti hai.

Syntax:

```python
iterator = iter(iterable)
```

---

### next() Function

- `next()` iterator ki next value return karta hai.
- Har baar call karne par next element milta hai.
- Jab saari values finish ho jati hain to `StopIteration` error aata hai.

---

### Looping Through an Iterator

- `for` loop automatically:
  - `iter()`
  - `next()`

  ko handle karta hai.

- Isliye normally hume manually `next()` use nahi karna padta.

---

### Create Your Own Iterator

Custom iterator banane ke liye class mein do special methods likhte hain.

#### `__iter__()`

- Iterator object return karta hai.

#### `__next__()`

- Har call par next value return karta hai.
- Jab values finish ho jayein to `StopIteration` raise karna hota hai.

---

### StopIteration

- Iterator ki saari values complete hone ke baad Python `StopIteration` exception raise karta hai.
- Ye batata hai ke iterator mein ab koi next value nahi bachi.

---

### Best Practice

- Normal loops ke liye `for` loop use karein.
- `next()` tab use karein jab manually values control karni hon.
- Custom iterators tab banayein jab apni iteration logic create karni ho.

````python
# ==========================
# Iterable Example
# ==========================

# List ek iterable hai
fruits = ["Apple", "Mango", "Banana"]


# ==========================
# Create Iterator
# ==========================

# List ko iterator mein convert kiya
fruit_iterator = iter(fruits)


# ==========================
# next() Function
# ==========================

# Har baar next() next value return karega
print(next(fruit_iterator))   # Apple

print(next(fruit_iterator))   # Mango

print(next(fruit_iterator))   # Banana

# Agar dobara next() call karoge
# to StopIteration error aayega

# print(next(fruit_iterator))


print("----------------")


# ==========================
# String Iterator
# ==========================

name = "Kainat"

letters = iter(name)

print(next(letters))
print(next(letters))
print(next(letters))


print("----------------")


# ==========================
# for Loop
# ==========================

# for loop automatically iterator use karta hai

numbers = [10, 20, 30]

for num in numbers:

    print(num)


print("----------------")


# ==========================
# Custom Iterator
# ==========================

class MyNumbers:

    def __iter__(self):

        # Starting value
        self.num = 1

        return self

    def __next__(self):

        # Jab number 5 se zyada ho jaye
        # to iteration stop kar do
        if self.num > 5:
            raise StopIteration

        current = self.num

        self.num += 1

        return current


# Object create kiya
my_object = MyNumbers()

# Iterator banaya
my_iterator = iter(my_object)

# Values print hongi
for value in my_iterator:

    print(value)
````

## Python Modules

### What is a Module?

- Module ek Python file (`.py`) hoti hai jisme functions, variables ya classes hoti hain.
- Module ka purpose code ko organize aur reusable banana hai.
- Ek baar module bana lo to usay multiple projects mein use kar sakte ho.

Example:

```
my_module.py
```

Ye ek module hai.

---

### Create a Module

- Module banane ke liye sirf ek Python file create karo.
- Us file mein variables, functions ya classes likh sakte ho.
- File save hote hi wo module ban jati hai.

---

### Use a Module

- Kisi module ko use karne ke liye `import` keyword use hota hai.

Syntax:

```python
import module_name
```

- Import karne ke baad us module ke functions aur variables use kar sakte hain.

---

### Module with Alias

- Agar module ka naam lamba ho to `as` keyword se short name de sakte hain.
- Isay **Alias** kehte hain.

Syntax:

```python
import module_name as short_name
```

Example:

```python
import math as m
```

Ab `math` ki jagah `m` use kar sakte hain.

---

### Built-in Modules

- Python mein bohat se modules pehle se available hote hain.
- Inhein **Built-in Modules** kehte hain.

Examples:

- math
- random
- datetime
- os

Inhein install karne ki zarurat nahi hoti.

---

### Using the dir() Function

- `dir()` kisi module ke andar available functions, variables aur methods ki list dikhata hai.

Syntax:

```python
dir(module_name)
```

---

### Import Only Specific Items

- Agar poora module import nahi karna to sirf required function ya variable import kar sakte hain.

Syntax:

```python
from module_name import function_name
```

Example:

```python
from math import sqrt
```

Ab `math.sqrt()` likhne ki zarurat nahi.

Seedha:

```python
sqrt(25)
```

likh sakte hain.

---

### Best Practice

- Sirf zarurat wala module import karein.
- Agar sirf ek function chahiye ho to `from ... import ...` use karein.
- Alias tab use karein jab module ka naam lamba ho.
- Apne modules ke meaningful names rakhein.

````python
# ==========================
# 1. Create Your Own Module
# Sabse pehle ek file banao:
# my_module.py
# Us file ke andar ye code likho:

# ---------------- my_module.py ----------------

# student_name = "Kainat"

# def welcome():
#     print("Welcome to Python Modules!")

# # ----------------------------------------------


# ==========================
# 2. Import Your Module
# ==========================

# # Main file mein likho

# import my_module

# # Module ka variable access karna
# print(my_module.student_name)

# # Module ka function call karna
# my_module.welcome()


# print("------------------------")


# ==========================
# 3. Module Alias
# ==========================

import math as m

# m ki wajah se math ki jagah short name use ho raha hai
print(m.sqrt(25))


print("------------------------")


# ==========================
# 4. Built-in Modules
# ==========================

import random

# Random number generate karega
print(random.randint(1, 10))


print("------------------------")


# ==========================
# 5. Import Only One Function
# ==========================

from math import sqrt

# Ab math.sqrt() likhne ki zarurat nahi
print(sqrt(49))


print("------------------------")


# ==========================
# 6. dir() Function
# ==========================

import math

# math module ke andar available sab functions ki list
print(dir(math))
````

## Python Dates (datetime)

### What is datetime?

- Python mein Date aur Time handle karne ke liye **datetime** module use hota hai.
- Ye built-in module hai, isliye install karne ki zarurat nahi hoti.
- Is module ki madad se current date, current time aur specific date/time ke sath kaam kar sakte hain.

---

### Import datetime Module

- `datetime` use karne se pehle module import karna zaroori hai.

Syntax:

```python
import datetime
```

---

### Current Date and Time

- Current date aur current time lene ke liye `datetime.now()` use hota hai.

Syntax:

```python
datetime.datetime.now()
```

Ye current system ki date aur time return karta hai.

---

### Access Individual Parts

Current date/time se alag alag values bhi nikal sakte hain.

Examples:

- Year
- Month
- Day
- Hour
- Minute
- Second

Ye properties use karke access hoti hain.

Example:

```python
today.year
today.month
today.day
```

---

### Create a Specific Date

- Agar apni koi custom date banani ho to `datetime()` constructor use karte hain.

Syntax:

```python
datetime.datetime(year, month, day)
```

Time bhi add kar sakte hain.

Syntax:

```python
datetime.datetime(year, month, day, hour, minute, second)
```

---

### strftime() Method

- `strftime()` ka matlab hai **String Format Time**.
- Iska use date ko apni pasand ke format mein show karne ke liye hota hai.

Syntax:

```python
date.strftime("format code")
```

---

### Common Format Codes

| Code | Meaning |
|------|---------|
| %Y | Full Year (2026) |
| %y | Short Year (26) |
| %m | Month Number |
| %B | Full Month Name |
| %b | Short Month Name |
| %d | Day |
| %A | Full Day Name |
| %a | Short Day Name |
| %H | Hour (24-hour) |
| %I | Hour (12-hour) |
| %M | Minute |
| %S | Second |
| %p | AM / PM |

---

### Useful Examples

Format:

```python
"%d/%m/%Y"
```

Output:

```
19/07/2026
```

---

Format:

```python
"%A"
```

Output:

```
Sunday
```

---

Format:

```python
"%B"
```

Output:

```
July
```

---

### Why Use datetime?

- Current date show karne ke liye.
- Current time show karne ke liye.
- Age calculate karne ke liye.
- Attendance systems.
- Billing systems.
- Booking systems.
- Reports generate karne ke liye.

---

### Best Practice

- Date aur Time ke liye hamesha `datetime` module use karein.
- User ko readable date dikhane ke liye `strftime()` use karein.
- Different countries ke format ke hisab se date display kar sakte hain.

````python
# ==========================
# Import datetime Module
# ==========================

import datetime


# ==========================
# Current Date & Time
# ==========================

# Abhi ki current date aur time store ho rahi hai
today = datetime.datetime.now()

print(today)


print("------------------------")


# ==========================
# Access Individual Parts
# ==========================

# Sirf year print hoga
print("Year :", today.year)

# Sirf month number print hoga
print("Month :", today.month)

# Sirf day print hoga
print("Day :", today.day)

# Current hour
print("Hour :", today.hour)

# Current minute
print("Minute :", today.minute)

# Current second
print("Second :", today.second)


print("------------------------")


# ==========================
# Create Your Own Date
# ==========================

# Apni custom date bana rahe hain

birthday = datetime.datetime(2004, 3, 11)

print(birthday)


print("------------------------")


# ==========================
# Create Date & Time Together
# ==========================

meeting = datetime.datetime(2026, 8, 5, 10, 30, 0)

print(meeting)


print("------------------------")


# ==========================
# strftime()
# ==========================

# Full year
print(today.strftime("%Y"))

# Month name
print(today.strftime("%B"))

# Day name
print(today.strftime("%A"))

# Date format
print(today.strftime("%d/%m/%Y"))

# Time format
print(today.strftime("%H:%M:%S"))


print("------------------------")


# ==========================
# More Format Examples
# ==========================

print("Short Year :", today.strftime("%y"))

print("Short Month :", today.strftime("%b"))

print("Short Day :", today.strftime("%a"))

print("AM/PM :", today.strftime("%p"))
````
