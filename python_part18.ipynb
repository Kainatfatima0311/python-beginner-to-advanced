## Matplotlib - Introduction

### What is Matplotlib?

- Matplotlib is a Python library for data visualization.
- It is used to create graphs and charts.

### Install

```bash
pip install matplotlib
```

### Import

```python
import matplotlib.pyplot as plt
```

### Create a Line Graph

```python
plt.plot(x, y)
plt.show()
```

### Uses

- Data Analysis
- Machine Learning
- Reports
- Dashboards
- Scientific Research

### Best Practice

- Import `matplotlib.pyplot` as `plt`.
- Always use `plt.show()` to display the graph.

````python
import matplotlib.pyplot as plt

# ==========================================
# Example 1: Simple Line Graph
# ==========================================

x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

plt.plot(x, y)
plt.show()
````

### Best Practice

- Always add axis labels.
- Give your graph a meaningful title.
- Use `fontdict` only when you need custom styling.

````python
import matplotlib.pyplot as plt

# Sample Data
months = ["Jan", "Feb", "Mar", "Apr"]
sales = [120, 150, 180, 200]

# Font Style
font = {
    "family": "serif",
    "color": "orange",
    "size": 16,
    "weight": "bold"
}

# Plot Graph
plt.plot(months, sales, marker="o")

# Labels
plt.xlabel("Months")
plt.ylabel("Sales")

# Title
plt.title("Monthly Sales Report", fontdict=font, loc="center")

# Show Graph
plt.show()
````

## Matplotlib - Subplot

### What is a Subplot?

- A subplot displays multiple graphs in the same figure.

### Syntax

```python
plt.subplot(rows, columns, position)
```

### Example

```python
plt.subplot(1,2,1)
```

Meaning:

- 1 Row
- 2 Columns
- Position 1

### 2×2 Layout

```text
1  2
3  4
```

### Individual Title

```python
plt.title("Sales")
```

### Overall Figure Title

```python
plt.suptitle("Company Report")
```

### Best Practice

- Use subplots when comparing multiple datasets.
- Give each subplot a meaningful title.
- Use `suptitle()` for the complete figure.

````python
import matplotlib.pyplot as plt

# Sample Data
months = ["Jan", "Feb", "Mar", "Apr"]

sales = [120, 150, 180, 200]
profit = [40, 55, 70, 90]
expenses = [80, 95, 110, 120]
customers = [300, 340, 390, 450]

# ==========================
# Sales
# ==========================
plt.subplot(2, 2, 1)
plt.plot(months, sales, marker="o")
plt.title("Sales")

# ==========================
# Profit
# ==========================
plt.subplot(2, 2, 2)
plt.plot(months, profit, marker="o")
plt.title("Profit")
plt.subplots_adjust(hspace=0.5)

# ==========================
# Expenses
# ==========================
plt.subplot(2, 2, 3)
plt.plot(months, expenses, marker="o")
plt.title("Expenses",loc = "left")

# ==========================
# Customers
# ==========================
plt.subplot(2, 2, 4)
plt.plot(months, customers, marker="o")
plt.title("Customers",loc= "right")

# Overall Title
plt.suptitle("Company Performance Report")

# Display Figure
plt.show()
````

## Matplotlib - Scatter Plot

### What is a Scatter Plot?

- Displays data as individual points.
- Points are not connected with lines.
- Used to find relationships between variables.

### Syntax

```python
plt.scatter(x, y)
```

### Color

```python
color="red"
```

### Size

```python
s=100
```

### Multiple Colors

```python
c=colors
```

### Colormap

```python
c=color_values
cmap="viridis"

plt.colorbar()
```

### Transparency

```python
alpha=0.5
```

### Marker

```python
marker="*"
```

### Best Practice

- Use scatter plots to compare two numeric variables.
- Use `alpha` when points overlap.
- Use `colorbar()` with `cmap` for numeric color values.

````python
import matplotlib.pyplot as plt

# Sample Data
hours = [1, 2, 3, 4, 5]
marks = [45, 55, 65, 75, 90]

# ==========================
# Basic Scatter Plot
# ==========================
plt.scatter(hours, marks)
plt.show()

````

````python
# ==========================
# Color & Size
# ==========================
plt.scatter(
    hours,
    marks,
    color="green",
    s=120
)
plt.show()

````

````python
# ==========================
# Different Colors
# ==========================
colors = ["red", "blue", "green", "orange", "purple"]

plt.scatter(
    hours,
    marks,
    c=colors,
    s=120
)
plt.show()

````

````python
# ==========================
# Colormap
# ==========================
values = [10, 20, 30, 40, 50]

plt.scatter(
    hours,
    marks,
    c=values,
    cmap="viridis",
    s=120
)

plt.colorbar()

plt.show()

````

````python
# ==========================
# Transparency
# ==========================
plt.scatter(
    hours,
    marks,
    color="blue",
    alpha=0.5,
    s=120
)

plt.show()
````

## Matplotlib - Histograms

### What is a Histogram?

- Displays the distribution of numerical data.
- Groups values into intervals called bins.

### Syntax

```python
plt.hist(data)
```

### Custom Bins

```python
bins=5
```

### Color

```python
color="green"
```

### Border

```python
edgecolor="black"
```

### Random Data

```python
import numpy as np

data = np.random.normal(
    170,
    10,
    250
)
```

### Histogram vs Bar Chart

Histogram

- Numerical data
- Distribution
- Uses bins

Bar Chart

- Categories
- Comparison
- Uses labels

### Best Practice

- Use histograms for continuous numerical data.
- Add `edgecolor` to make bins easy to distinguish.

````python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# Example 1: Simple Histogram
# ==========================================

marks = [45, 50, 60, 65, 70, 72, 75, 80, 82, 90]

plt.hist(marks)
plt.title("Simple Histogram")
plt.show()

````

````python
# ==========================================
# Example 2: Custom Bins
# ==========================================

plt.hist(
    marks,
    bins=5,
    color="skyblue",
    edgecolor="black"
)

plt.title("Histogram with Custom Bins")
plt.show()

````

````python
# ==========================================
# Example 3: Random Data
# ==========================================

data = np.random.normal(
    170,
    10,
    250
)

plt.hist(
    data,
    bins=10,
    color="orange",
    edgecolor="black"
)

plt.title("Random Height Distribution")
plt.xlabel("Height")
plt.ylabel("Frequency")

plt.show()
````

## Matplotlib - Pie Charts

### What is a Pie Chart?

- Displays data as slices of a circle.
- Shows how each category contributes to the whole.

### Syntax

```python
plt.pie(data)
```

### Labels

```python
labels=labels
```

### Percentage

```python
autopct="%1.1f%%"
```

### Rotate Chart

```python
startangle=90
```

### Explode

```python
explode=[0.1,0,0,0]
```

### Colors

```python
colors=["red","green","blue","orange"]
```

### Shadow

```python
shadow=True
```

### Legend

```python
plt.legend()
```

### Best Practice

- Use pie charts only when showing parts of a whole.
- Avoid too many slices.
- Use `autopct` to display percentages.

````python
import matplotlib.pyplot as plt

# Sample Data
budget = [40, 25, 20, 15]

labels = [
    "Food",
    "Rent",
    "Transport",
    "Entertainment"
]

colors = [
    "gold",
    "skyblue",
    "lightgreen",
    "orange"
]

explode = [0.1, 0, 0, 0]

# Create Pie Chart
plt.pie(
    budget,
    labels=labels,
    colors=colors,
    explode=explode,
    autopct="%1.1f%%",
    shadow=True,
    startangle=90
)

plt.title("Monthly Budget")
plt.legend(title="Categories")

plt.show()
````
