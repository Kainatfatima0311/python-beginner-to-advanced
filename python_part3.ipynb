#### **Python Data Types**

A **data type** tells us what kind of value a variable stores.

| **Data Type** | **Meaning** | **Example** |
|---------------|-------------|-------------|
| `int` | Whole numbers | `5` |
| `float` | Decimal numbers | `5.5` |
| `complex` | Complex numbers | `3 + 4j` |
| `str` | Text (String) | `"Hello"` |
| `bool` | True or False | `True` |
| `list` | Ordered, changeable collection | `[1, 2, 3]` |
| `tuple` | Ordered, unchangeable collection | `(1, 2, 3)` |
| `range` | Sequence of numbers | `range(5)` |
| `dict` | Key-value pairs | `{"name": "Ali"}` |
| `set` | Unordered collection of unique values | `{1, 2, 3}` |
| `frozenset` | Read-only set | `frozenset({1, 2})` |
| `bytes` | Immutable byte data | `b"Hello"` |
| `bytearray` | Mutable byte data | `bytearray(5)` |
| `memoryview` | Memory view of binary data | `memoryview(bytes(5))` |
| `NoneType` | Represents no value | `None` |

````python
price = 99.99

print(type(price))
````

#### **Python Numbers**

Python has **three numeric data types**.

- **`int`** → Stores whole numbers.
- **`float`** → Stores decimal numbers.
- **`complex`** → Stores complex numbers (real + imaginary).

````python
x=3
y=7.8
z=5j
print(type(x))
print(type(y))
print(type(z))
````

#### **Python Type Conversion**

Type Conversion means **changing one data type into another**.

There are **two types** of type conversion:

##### **Implicit Type Conversion**
- Python automatically converts the data type when needed.

##### **Explicit Type Conversion (Type Casting)**
- The programmer manually converts the data type using functions like `int()`, `float()`, or `str()`.

````python
#implicit type casting
x=2
y=4.5
sum=x+y
print(sum)
print(type(x))
print(type(y))
print(type(sum))
````

````python
#explicit type casting
x=5.6
x=int(x)
print(type(x))
print(x)
c = complex(x)
print(type(c))
print(c)
````

#### **Python Random Number**

A **random number** is a number that can be different each time the program runs.

- Python uses the **`random`** module to generate random numbers.
- First, import the `random` module before using its functions.
- The `random` module provides different functions to generate random values.

````python
import random
print(random.randrange(1,10))
````

````python
import random
x=int(input("Enter guess number between 1 to 10:"))
y=random.randrange(1,10)

if x==y:
  print("Congrats!your guess is right")
else:
  print("Oops! Failed")
print("Its actually..", y)
````

- Strings are used to store names, messages, sentences, and other text.

````python
x="kf"
y='kf'
print(x)
print(y)

#qoute mn agr number bhi tu wo str h
````

#### **Multi-line Strings**

Use **triple quotes** (`''' '''` or `""" """`) to write a string on multiple lines.

- Triple quotes preserve line breaks.
- Useful for long text or paragraphs.

````python
text = """
Hello
Welcome to Python
Have a nice day!
"""

print(text)
````

#### **String is a Collection**

A string is a **collection (sequence) of characters**.

- Each character has its own **index**.
- You can access characters using their index position.
- A string can contain **letters, numbers, spaces, and special symbols**.

````python
txt="message"
print(txt[3])

#length

print(len(txt))
````

#### **Looping Through a String**

You can use a **`for` loop** to access each character of a string one by one.

**Syntax:**

```text
for variable in string:
    print(variable)
```

````python
for letter in "Python":
    print(letter)
````

#### **Check String**

Use the **`in`** and **`not in`** operators to check whether a character or word exists in a string.

- **`in`** → Returns `True` if the value exists.
- **`not in`** → Returns `True` if the value does not exist.

````python
x="Life is so unpridictable"
print("is" in x)
````

````python
x="Life is so unpridictable"
print("is" not  in x)
````
