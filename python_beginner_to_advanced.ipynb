#### **Variable Naming Rules**

- A variable name **must start with a letter** (`a-z`, `A-Z`).
- It can also **start with an underscore (`_`)**.
- A **number cannot be the first character**, but it can come later.
- **Spaces are not allowed** in variable names.
- **Special characters** (such as `@`, `#`, `$`, `%`) are **not allowed** except the underscore (`_`).
- **Python keywords** (such as `if`, `for`, `while`, `class`) **cannot** be used as variable names.

#### **Operator Precedence**

Operator precedence decides **which operation is performed first** when an expression contains multiple operators.

| **Priority** | **Operator** |
|--------------|--------------|
| 1 | `()` |
| 2 | `**` |
| 3 | `*`, `/`, `//`, `%` |
| 4 | `+`, `-` |
| 5 | `==`, `!=`, `>`, `<`, `>=`, `<=` |
| 6 | `not` |
| 7 | `and` |
| 8 | `or` |

````python
# ==========================
# Arithmetic Operators
# ==========================

a = 10
b = 5

# Addition
print(a + b)

# Subtraction
print(a - b)

# Multiplication
print(a * b)

# Division
print(a / b)

# Floor Division (Removes the decimal part)
print(a // b)

# Modulus (Returns the remainder)
print(a % b)

# Exponent (Power)
print(a ** b)


# ==========================
# Assignment Operators
# ==========================

x = 10
print("Initial Value:", x)

x += 5    # x = x + 5
print("After += :", x)

x -= 10   # x = x - 10
print("After -= :", x)

x *= 20   # x = x * 20
print("After *= :", x)


# ==========================
# Comparison Operators
# ==========================

c = 10
d = 7

print(c == d)   # Equal
print(c != d)   # Not Equal
print(c > d)    # Greater Than
print(c < d)    # Less Than
print(c >= d)   # Greater Than or Equal
print(c <= d)   # Less Than or Equal


# ==========================
# Logical Operators
# ==========================

# AND Operator (Both conditions must be True)
age = 19
student = True

print(age >= 18 and student)

# OR Operator (At least one condition must be True)
print(age >= 18 or student)

# NOT Operator (Reverses the result)
is_login = True

print(not is_login)
````

````python
# ==========================
# Identity Operators
# ==========================
# Identity operators compare the memory location (object identity).

a = [1, 2, 3]
b = a

print(a is b)      # True (Both refer to the same object)

b = [1, 2, 3]

print(a == b)      # True (Values are the same)
print(a is b)      # False (Different objects in memory)


# ==========================
# Membership Operators
# ==========================
# Check whether a value exists in a collection.

fruits = ["Apple", "Mango", "Banana"]

print("Mango" in fruits)        # True
print("Orange" not in fruits)   # True


# ==========================
# Bitwise Operators
# ==========================
# Bitwise operators work on binary (0 and 1) values.

a = 5   # Binary: 101
b = 3   # Binary: 011

print(a & b)   # AND
print(a | b)   # OR
print(a ^ b)   # XOR
print(~a)      # NOT
print(a << 1)  # Left Shift
print(a >> 1)  # Right Shift


# ==========================
# Ternary Operator
# ==========================
# value_if_true if condition else value_if_false

age = 21

result = "Adult" if age >= 18 else "Not Adult"

print(result)
````
