## Python JSON

### What is JSON?

- **JSON** ka full form **JavaScript Object Notation** hai.
- Ye data store aur data exchange karne ka ek common format hai.
- JSON ko APIs aur web applications mein bohat use kiya jata hai.
- Python mein JSON data ko handle karne ke liye **json** module use hota hai.

Example JSON:

```json
{
  "name": "Kainat",
  "age": 22,
  "city": "Lahore"
}
```

---

### Import JSON Module

- JSON use karne ke liye pehle module import karna hota hai.

Syntax:

```python
import json
```

---

### Python Object to JSON (json.dumps())

- `dumps()` ka matlab **Dump String** hai.
- Ye Python object ko JSON String mein convert karta hai.

Syntax:

```python
json.dumps(python_object)
```

---

### JSON to Python Object (json.loads())

- `loads()` ka matlab **Load String** hai.
- Ye JSON String ko Python object mein convert karta hai.

Syntax:

```python
json.loads(json_string)
```

---

### Python Data Types Converted into JSON

| Python | JSON |
|---------|------|
| dict | Object |
| list | Array |
| tuple | Array |
| str | String |
| int | Number |
| float | Number |
| True | true |
| False | false |
| None | null |

---

### Formatting JSON

`json.dumps()` mein formatting options bhi use kar sakte hain.

---

### indent

- JSON ko readable banata hai.
- Har level par spacing add karta hai.

Example:

```python
json.dumps(data, indent=4)
```

---

### separators

- JSON mein comma aur colon ka format change karta hai.

Example:

```python
json.dumps(data, separators=(".", "="))
```

---

### sort_keys

- Dictionary keys ko alphabetical order mein arrange karta hai.

Example:

```python
json.dumps(data, sort_keys=True)
```

---

### Why Use JSON?

JSON ka use:

- APIs
- Web Applications
- Mobile Apps
- Data Storage
- Client aur Server ke darmiyan data transfer

mein hota hai.

---

### Best Practice

- JSON data ko handle karne ke liye hamesha `json` module use karein.
- Readable output ke liye `indent` use karein.
- Large projects mein JSON APIs ke sath bohat use hota hai.

````python
# ==========================
# Import JSON Module
# ==========================

import json


# ==========================
# Python Dictionary
# ==========================

# Ye normal Python dictionary hai
student = {
    "name": "Kainat",
    "age": 22,
    "city": "Lahore",
    "is_student": True
}

print(student)


print("------------------------")


# ==========================
# Python Object -> JSON
# json.dumps()
# ==========================

# Dictionary ko JSON String mein convert kiya
json_data = json.dumps(student)

print(json_data)

# Data type check
print(type(json_data))
# Output: <class 'str'>


print("------------------------")


# ==========================
# JSON -> Python Object
# json.loads()
# ==========================

# JSON String ko dobara Python Dictionary banaya
python_data = json.loads(json_data)

print(python_data)

# Data type check
print(type(python_data))
# Output: <class 'dict'>


print("------------------------")


# ==========================
# indent
# ==========================

# Pretty format mein JSON print hoga
print(json.dumps(student, indent=4))


print("------------------------")


# ==========================
# sort_keys
# ==========================

# Keys alphabetical order mein print hongi
print(json.dumps(student, indent=4, sort_keys=True))


print("------------------------")


# ==========================
# separators
# ==========================

# Separator change kiya
print(json.dumps(student, separators=(" | ", " = ")))


print("------------------------")


# ==========================
# Different Python Objects
# ==========================

print(json.dumps(["Apple", "Mango", "Banana"]))   # List

print(json.dumps(("Red", "Blue")))                # Tuple

print(json.dumps("Hello"))                        # String

print(json.dumps(100))                            # Integer

print(json.dumps(25.5))                           # Float

print(json.dumps(True))                           # Boolean

print(json.dumps(None))                           # None
````

## Python RegEx

### What is RegEx?

- **RegEx** ka full form **Regular Expression** hai.
- RegEx ka use text ya string mein specific pattern ko search, check ya replace karne ke liye hota hai.
- Python mein RegEx use karne ke liye **re** module import karte hain.

Example:

- Email check karna
- Phone number validate karna
- Password validation
- Word search karna

---

### Import re Module

- RegEx use karne se pehle `re` module import karna zaroori hai.

Syntax:

```python
import re
```

---

### RegEx Functions

Python mein mostly ye functions use hote hain:

- `findall()`
- `search()`
- `split()`
- `sub()`

---

### findall()

- String mein jitni matching values hoti hain un sab ki list return karta hai.
- Agar match na mile to empty list `[]` return hoti hai.

Syntax:

```python
re.findall(pattern, string)
```

---

### search()

- String mein pehli matching value dhoondta hai.
- Match mil jaye to Match Object return karta hai.
- Match na mile to `None` return karta hai.

Syntax:

```python
re.search(pattern, string)
```

---

### split()

- Pattern milne par string ko split karta hai.
- Result list ki form mein milta hai.

Syntax:

```python
re.split(pattern, string)
```

---

### sub()

- Pattern ko kisi doosri value se replace karta hai.

Syntax:

```python
re.sub(pattern, replace_value, string)
```

---

## Match Object

Agar `search()` ko match mil jaye to Match Object return hota hai.

Useful Methods:

- `group()`
- `start()`
- `end()`
- `span()`

Ye batate hain:

- Match kya mila.
- Kahan se start hua.
- Kahan end hua.

---

## Best Practice

- Email validation.
- Phone number validation.
- Password validation.
- Search & Replace.
- Text Processing.

````python
# ==========================
# Import re Module
# ==========================

import re


text = "My name is Kainat. My roll number is 12345."


# ==========================
# findall()
# ==========================

# "My" jitni baar milega sab list mein aa jayega
result = re.findall("My", text)

print(result)


print("------------------")


# ==========================
# search()
# ==========================

# Pehla match dhoondta hai
result = re.search("Kainat", text)

# group() matched word return karta hai
print(result.group())

# Match kis index se start hua
print(result.start())

# Match kis index par end hua
print(result.end())

# Start aur End dono ek sath
print(result.span())


print("------------------")


# ==========================
# split()
# ==========================

# Space ki jaga string split hogi
result = re.split(" ", text)

print(result)


print("------------------")


# ==========================
# sub()
# ==========================

# "Kainat" ko "Ali" se replace kar diya
result = re.sub("Kainat", "Ali", text)

print(result)


print("------------------")


# ==========================
# \d  (Digits)
# ==========================

# Sirf digits find karega
result = re.findall(r"\d", text)

print(result)


print("------------------")


# ==========================
# \w (Letters + Digits)
# ==========================

result = re.findall(r"\w", text)

print(result[:15])   # Sirf first 15 values print ki


print("------------------")


# ==========================
# [a-z]
# ==========================

# Small letters
result = re.findall(r"[a-z]", text)

print(result[:20])


print("------------------")


# ==========================
# ^
# ==========================

text2 = "Python is Easy"

result = re.findall(r"^Python", text2)

print(result)


print("------------------")


# ==========================
# $
# ==========================

result = re.findall(r"Easy$", text2)

print(result)


print("------------------")


# ==========================
# |
# ==========================

text3 = "I have a cat and a dog"

result = re.findall(r"cat|dog", text3)

print(result)
````

## Python Try...Except

### What is Exception Handling?

- Jab program run hote waqt koi error aata hai to usay **Exception** kehte hain.
- Agar exception handle na ki jaye to program wahi stop ho jata hai.
- `try...except` ki madad se error handle kar sakte hain aur program crash nahi hota.

---

### try Block

- Jis code mein error aane ka chance ho usay `try` block mein likhte hain.
- Python pehle `try` block execute karta hai.

Syntax:

```python
try:
    # Code
```

---

### except Block

- Agar `try` block mein error aa jaye to `except` block execute hota hai.
- Is block mein error ko handle karte hain.

Syntax:

```python
except:
    # Error Handle
```

---

### Many Exceptions

- Different errors ke liye alag alag `except` blocks likh sakte hain.
- Is se har error ko alag tareeqe se handle kiya ja sakta hai.

Example:

```python
except NameError:
except TypeError:
except ZeroDivisionError:
```

---

### else Block

- `else` tab execute hota hai jab `try` block mein **koi error na aaye**.
- Agar error aa jaye to `else` execute nahi hota.

---

### finally Block

- `finally` hamesha execute hota hai.
- Chahe error aaye ya na aaye.
- Iska use cleanup work ke liye hota hai.

Examples:

- File close karna.
- Database connection close karna.

---

### Raise an Exception

- Kabhi kabhi hum khud bhi error generate karna chahte hain.
- Iske liye `raise` keyword use hota hai.

Syntax:

```python
raise Exception("Message")
```

Ya

```python
raise ValueError("Message")
```

---

### Common Exceptions

| Exception | Reason |
|-----------|--------|
| NameError | Variable exist nahi karta |
| TypeError | Wrong data type |
| ValueError | Invalid value |
| IndexError | Wrong index |
| KeyError | Dictionary key exist nahi karti |
| ZeroDivisionError | Number ko 0 se divide karna |
| FileNotFoundError | File na milna |

---

### Why Use try...except?

- Program crash hone se bach jata hai.
- User-friendly error messages dikhaye ja sakte hain.
- Program smoothly continue karta hai.

---

### Best Practice

- Sirf us code ko `try` mein rakhein jahan error ka chance ho.
- Specific exception use karein instead of simple `except`.
- `finally` cleanup work ke liye use karein.
- Har error ko silently ignore na karein.

````python
# ==========================================
# 1. Simple try...except
# ==========================================

try:
    # 10 ko 0 se divide nahi kar sakte
    print(10 / 0)

except:
    # Error aane par ye chalega
    print("Error Handle Ho Gaya")


print("----------------------")


# ==========================================
# 2. Specific Exception
# ==========================================

try:
    print(10 / 0)

except ZeroDivisionError:
    # Sirf ZeroDivisionError ko handle karega
    print("Number ko 0 se divide nahi kar sakte.")


print("----------------------")


# ==========================================
# 3. Multiple Exceptions
# ==========================================

try:

    number = int("Hello")

except ValueError:

    print("Invalid Number")

except TypeError:

    print("Type Error")


print("----------------------")


# ==========================================
# 4. else Block
# ==========================================

try:

    num1 = 10
    num2 = 2

    print(num1 / num2)

except ZeroDivisionError:

    print("Division Error")

else:

    # Sirf tab chalega jab koi error nahi aayega
    print("Program Successfully Run Ho Gaya")


print("----------------------")


# ==========================================
# 5. finally Block
# ==========================================

try:

    print("Try Block")

except:

    print("Error")

finally:

    # Ye hamesha execute hoga
    print("Finally Block Always Runs")


print("----------------------")


# ==========================================
# 6. raise Exception
# ==========================================

age = -5

try:

    if age < 0:

        # Khud apna error generate kiya
        raise ValueError("Age Negative Nahi Ho Sakti")

    print(age)

except ValueError as error:

    # Error message print karega
    print(error)
````

## Python String Formatting

### What is String Formatting?

- String Formatting ka matlab variables ki values ko string ke andar insert karna hai.
- Is se output zyada readable aur professional hoti hai.
- Python mein sab se common tareeqa **f-strings** hai.

Example:

```python
name = "Kainat"
print(f"Hello {name}")
```

Output:

```
Hello Kainat
```

---

## F-Strings

- **f-string** Python 3.6 se introduce hui.
- String ke start mein `f` ya `F` lagaya jata hai.
- Variable ya expression `{}` ke andar likhte hain.

Syntax:

```python
f"Text {variable}"
```

---

### Insert Variables

Example:

```python
name = "Ali"
age = 20

print(f"My name is {name} and I am {age} years old.")
```

---

### Insert Expressions

- `{}` ke andar calculations bhi kar sakte hain.

Example:

```python
price = 50
quantity = 4

print(f"Total = {price * quantity}")
```

---

## Format Specifiers

Format specifiers output ka format control karte hain.

Syntax:

```python
{value:specifier}
```

---

### :.2f

- Float ko **2 decimal places** tak show karta hai.

Example:

```python
price = 59.9876

print(f"{price:.2f}")
```

Output:

```
59.99
```

---

### :.3f

- Float ko **3 decimal places** tak show karta hai.

---

### :,

- Large numbers mein comma laga deta hai.

Example:

```python
number = 1000000

print(f"{number:,}")
```

Output:

```
1,000,000
```

---

### :%

- Decimal number ko percentage mein convert karta hai.

Example:

```python
marks = 0.85

print(f"{marks:%}")
```

Output:

```
85.000000%
```

---

### :.2%

- Percentage ko 2 decimal places ke sath show karta hai.

Example:

```python
print(f"{marks:.2%}")
```

Output:

```
85.00%
```

---

## Formatting Numbers

Number formatting ko readable banane ke liye use karte hain.

Examples:

- Decimal places
- Comma separator
- Percentage

---

## Why Use String Formatting?

- Output readable hoti hai.
- Printing easy ho jati hai.
- Variables ko text ke sath easily combine kar sakte hain.
- Reports aur user messages mein bohat use hota hai.

---

## Best Practice

- Modern Python mein **f-strings** use karein.
- Expressions directly `{}` ke andar likh sakte hain.
- Decimal values ke liye `:.2f` commonly use hota hai.
- Currency aur reports mein `:,` bohat useful hota hai.

````python
# ==========================================
# Simple f-string
# ==========================================

name = "Kainat"

print(f"Hello {name}")


print("------------------------")


# ==========================================
# Multiple Variables
# ==========================================

name = "Kainat"
age = 22

print(f"My name is {name} and I am {age} years old.")


print("------------------------")


# ==========================================
# Expressions
# ==========================================

price = 500
quantity = 3

# Expression directly {}
print(f"Total Price = {price * quantity}")


print("------------------------")


# ==========================================
# Float Formatting
# ==========================================

number = 56.789456

print(f"2 Decimal = {number:.2f}")

print(f"3 Decimal = {number:.3f}")


print("------------------------")


# ==========================================
# Comma Separator
# ==========================================

salary = 2500000

print(f"Salary = {salary:,}")


print("------------------------")


# ==========================================
# Percentage
# ==========================================

marks = 0.875

print(f"Percentage = {marks:%}")

print(f"Percentage = {marks:.2%}")


print("------------------------")


# ==========================================
# Multiple Expressions
# ==========================================

num1 = 20
num2 = 10

print(f"Addition = {num1 + num2}")

print(f"Subtraction = {num1 - num2}")

print(f"Multiplication = {num1 * num2}")

print(f"Division = {num1 / num2:.2f}")
````

## Python None

### What is None?

- **None** Python mein ek special value hai.
- Ye represent karti hai ke **koi value nahi hai**.
- `None` koi string (`"None"`), number ya Boolean nahi hota.
- Iska data type `NoneType` hota hai.

Example:

```python
x = None
```

---

### NoneType

- `None` ka apna data type hota hai.

Example:

```python
x = None

print(type(x))
```

Output:

```
<class 'NoneType'>
```

---

### Checking for None

- `None` check karne ke liye **`is`** operator use karna best practice hai.

Syntax:

```python
variable is None
```

Example:

```python
name = None

if name is None:
    print("No Value")
```

---

### Why use `is` instead of `==`?

- `None` Python mein sirf ek hi special object hota hai.
- Isliye `is None` use karna recommended hai.

Best Practice:

```python
if value is None:
```

Not:

```python
if value == None:
```

---

### None in Functions

- Agar function `return` statement use nahi karta, to woh automatically `None` return karta hai.

Example:

```python
def hello():
    print("Hello")

result = hello()

print(result)
```

Output:

```
Hello
None
```

---

### Assigning None

Kabhi kabhi variable ki value baad mein deni hoti hai.

Example:

```python
user = None
```

Baad mein:

```python
user = "Kainat"
```

---

### None vs False vs 0 vs Empty String

| Value | Meaning |
|--------|---------|
| None | No value |
| False | Boolean false |
| 0 | Number zero |
| "" | Empty string |
| [] | Empty list |

Ye sab alag alag values hain.

Example:

```python
None == False
```

Output:

```
False
```

---

### Where is None Used?

- Function return values
- Default function parameters
- Database records
- APIs
- Variable initialization

---

### Best Practice

- `None` check karne ke liye `is None` use karein.
- `"None"` (string) aur `None` ko confuse na karein.
- `None` ko missing value represent karne ke liye use karein.

````python
# ==========================================
# Creating a None Value
# ==========================================

value = None

print(value)

print(type(value))


print("------------------------")


# ==========================================
# Checking None
# ==========================================

name = None

if name is None:
    print("No Value Available")


print("------------------------")


# ==========================================
# None is not False
# ==========================================

print(None == False)

print(None == 0)

print(None == "")

print(None == [])


print("------------------------")


# ==========================================
# Using None in Variables
# ==========================================

student = None

print(student)

# Later value assign ki
student = "Kainat"

print(student)


print("------------------------")


# ==========================================
# Function Returning None
# ==========================================

def greet():
    print("Welcome to Python")


result = greet()

print(result)


print("------------------------")


# ==========================================
# Function Returning Value
# ==========================================

def add(num1, num2):
    return num1 + num2


answer = add(10, 20)

print(answer)


print("------------------------")


# ==========================================
# Comparing with is None
# ==========================================

data = None

if data is None:
    print("Data is Missing")
else:
    print("Data Found")
````

## Convert Input to Integer

Agar number par calculations karni hon to `int()` use karein.

Example:

```python
age = int(input("Enter your age: "))
```

---

## Convert Input to Float

Decimal numbers ke liye `float()` use hota hai.

Example:

```python
price = float(input("Enter price: "))
```

---

## Taking Multiple Inputs

Ek se zyada values bhi le sakte hain.

Example:

```python
name = input("Name: ")
age = int(input("Age: "))
```

---

## Performing Calculations

Input ko integer ya float mein convert karne ke baad calculations kar sakte hain.

Example:

```python
num1 = int(input("First Number: "))
num2 = int(input("Second Number: "))

print(num1 + num2)
```

---

## Why Type Conversion is Important?

Agar conversion na karein to numbers bhi string ki tarah treat honge.

Example:

```python
num1 = input()
num2 = input()

print(num1 + num2)
```

Input:

```
10
20
```

Output:

```
1020
```

Kyunkay dono strings hain.

---

## Best Practice

- Text ke liye direct `input()` use karein.
- Integer ke liye `int(input())`.
- Decimal ke liye `float(input())`.
- User input ko use karne se pehle validate karna real projects mein achi practice hai.

````python
# ==========================================
# 1. Taking String Input
# ==========================================

name = input("Enter your name: ")

print("Hello", name)


print("------------------------")


# ==========================================
# 2. Input is Always String
# ==========================================

age = input("Enter your age: ")

print(age)

print(type(age))


print("------------------------")


# ==========================================
# 3. Convert Input to Integer
# ==========================================

age = int(input("Enter your age: "))

print("Next Year Age:", age + 1)


print("------------------------")


# ==========================================
# 4. Convert Input to Float
# ==========================================

price = float(input("Enter Product Price: "))

print("Price:", price)


print("------------------------")


# ==========================================
# 5. Addition of Two Numbers
# ==========================================

num1 = int(input("Enter First Number: "))

num2 = int(input("Enter Second Number: "))

print("Sum =", num1 + num2)


print("------------------------")


# ==========================================
# 6. What Happens Without int()
# ==========================================

num1 = input("Enter First Number: ")

num2 = input("Enter Second Number: ")

# Ye string concatenate karega
print("Result =", num1 + num2)


print("------------------------")


# ==========================================
# 7. Multiple Inputs
# ==========================================

name = input("Enter Name: ")

city = input("Enter City: ")

print(f"{name} lives in {city}.")
````

## Python Virtual Environment

### What is a Virtual Environment?

- Virtual Environment ek **isolated Python environment** hota hai.
- Ismein har project ke apne packages aur dependencies hoti hain.
- Ek project ke packages doosre project ko affect nahi karte.

Example:

Project A

- pandas 2.2

Project B

- pandas 1.5

Dono projects apni apni version use kar sakte hain.

---

## Why Use Virtual Environment?

Without Virtual Environment:

- Sab packages global install hote hain.
- Version conflicts aa sakte hain.
- Ek project ki dependency doosre project ko break kar sakti hai.

With Virtual Environment:

- Har project ka apna environment.
- Har project ki apni dependencies.
- Projects independent rehte hain.

---

## Create a Virtual Environment

Command:

```bash
python -m venv myenv
```

- `venv` Python ka built-in module hai.
- `myenv` environment ka naam hai.
- Is command ke baad `myenv` naam ka folder ban jayega.

---

## Activate Virtual Environment

### Windows

Command:

```bash
myenv\Scripts\activate
```

---

### macOS / Linux

Command:

```bash
source myenv/bin/activate
```

---

## Activated Environment

Activate hone ke baad terminal mein environment ka naam show hota hai.

Example:

```text
(myenv) C:\Projects>
```

Iska matlab environment successfully activate ho gaya hai.

---

## Install Packages

Environment activate hone ke baad package install karein.

Example:

```bash
pip install camelcase
```

Ye package sirf isi virtual environment mein install hoga.

---

## View Installed Packages

Command:

```bash
pip list
```

Installed packages ki list show karega.

---

## Deactivate Virtual Environment

Kaam complete hone ke baad environment close karne ke liye:

```bash
deactivate
```

Terminal normal state mein aa jayegi.

---

## requirements.txt

Project ke installed packages save karne ke liye:

```bash
pip freeze > requirements.txt
```

Ye file project ki dependencies save karti hai.

---

## Install Packages from requirements.txt

Agar kisi aur system par project chalana ho to:

```bash
pip install -r requirements.txt
```

Ye sari required libraries automatically install kar dega.

---

## Why is requirements.txt Important?

- Team projects.
- Deployment.
- GitHub projects.
- Same package versions maintain karne ke liye.

---

## Best Practice

- Har naye project ke liye alag Virtual Environment banayein.
- `requirements.txt` GitHub par upload karein.
- Virtual Environment folder (`myenv`, `venv`, `.venv`) ko GitHub par upload na karein.
- `.gitignore` file mein Virtual Environment folder add karein.
