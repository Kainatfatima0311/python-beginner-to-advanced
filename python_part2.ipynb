#### **Input Variable**

The `input()` function is used to get input from the user.

- The value returned by `input()` is **always a string (`str`)**.
- If you need a number, convert it using `int()` or `float()`.

```python
name=input("Kindly enter your name: ")
print(name)
```

```python
age=input("enter your age: ")
print(age)

print(type(age))
```

#### **Dynamic Typing**

Python uses **dynamic typing**, which means you don't need to specify a variable's data type.

- Python automatically detects the data type based on the assigned value.
- A variable's type can change if a new value of a different type is assigned.
- This makes Python flexible and easy to use.

#### **Semicolons (`;`)**

Semicolons are **optional** in Python and are **rarely used**.

- They can be used to write **multiple statements on the same line**.
- Using semicolons makes the code **less readable**.
- It is recommended to write **one statement per line** for clean and easy-to-understand code.

```python
print('hi');print("! I'm trying to test ");print("use of semicolons")
```

#### **Print Without a New Line**

By default, each `print()` statement starts on a **new line**.

- Use the `end` parameter to print the next output on the **same line**.
- `end=" "` adds a space between the outputs.

```python
print("hi")
print("kainat", end=" ")
print("fatima")
```

#### **Mix Text and Numbers**

You can print **text and numbers** together in the same `print()` statement.

- Use a **comma (`,`)** to separate values.
- Python automatically adds a space between them.

```python
print("hi! I am ", 21 ," years old")
```

#### **Unpack a Collection**

Unpacking means **assigning the elements of a collection to multiple variables**.

- The number of variables must match the number of elements.
- It works with collections like **lists, tuples, and strings**.

```python
colours=["pink","red","black"]
a,b,c= colours
print(a)
print(b)
print(c)
```

#### **Global Variables**

A **global variable** is a variable created **outside a function**.

- It can be accessed from anywhere in the program, including inside functions.
- Global variables are available until the program ends.

```python
message = "Welcome"

def first():  #def ka mtlb define function
    print(message)

def second():
    print(message)

first()
second()
```
