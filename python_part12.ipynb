## File Handling

### What is File Handling?

- File Handling ka matlab files ko **create, open, read, write, update aur delete** karna hota hai.
- Iski help se data permanently save hota hai.
- Variables ka data program band hony par delete ho jata hai, lekin file ka data save rehta hai.
- File Handling real-world applications jese student records, reports, logs aur data storage me use hoti hai.

---

## Opening a File

- Python me file open karne ke liye `open()` function use hota hai.
- `open()` ko file ka naam aur mode diya jata hai.

### File Modes

- `r` → Read mode (Default)
- `w` → Write mode (Purana data overwrite kar deta hai.)
- `a` → Append mode (Naya data end me add karta hai.)
- `x` → Create mode (Sirf new file create karta hai.)
- `t` → Text mode (Default)
- `b` → Binary mode (Images, PDFs, etc.)

---

## Reading Files

- Read mode (`r`) file ka data read karne ke liye use hota hai.
- Agar file exist na kare to error aata hai.

### Reading Methods

- `read()` → Poori file read karta hai.
- `read(n)` → Sirf specified number of characters read karta hai.
- `readline()` → Sirf ek line read karta hai.
- File ko loop ke through line by line bhi read kiya ja sakta hai.

---

## Closing a File

- File par kaam complete hone ke baad usay close karna chahiye.
- `close()` system resources free karta hai.
- File ko close karna good programming practice hai.

---

## Writing Files

- Write mode (`w`) file me naya data likhta hai.
- Agar file already exist kare to uska purana data delete ho jata hai.
- Agar file exist na kare to new file create ho jati hai.

---

## Appending Data

- Append mode (`a`) existing data ko delete nahi karta.
- Naya data file ke end me add hota hai.

---

## Creating Files

- Create mode (`x`) sirf new file create karta hai.
- Agar file pehle se exist kare to error generate hota hai.

---

## Deleting Files

- File delete karne ke liye `os` module use hota hai.
- File delete karne se pehle check karna chahiye ke file exist karti hai ya nahi.
- Empty folder ko bhi delete kiya ja sakta hai.

---

## Important Points

- File ko open karne ke liye `open()` function use hota hai.
- `r` sirf file read karta hai.
- `w` puri file overwrite kar deta hai.
- `a` existing data ko preserve karke end me naya data add karta hai.
- `x` sirf new file create karta hai.
- `read()` complete file read karta hai.
- `readline()` ek line read karta hai.
- `close()` file ko safely close karta hai.
- File delete karne ke liye `os` module use hota hai.
- Empty folder hi delete kiya ja sakta hai.

```python
import os


#----------------------
print("Creating new file and write data")
#----------------------
print("----------------------")


file=open("demo.txt","w")
file.write("Hello python!\n")
file.write("Welcome to file handling")
file.close()

#----------------------
print("Reading complete file")
#----------------------

file=open("demo.txt","r")
print("complete file")
print(file.read())
file.close()

#----------------------
print("Reading line by line")
print("----------------------")
#----------------------

file=open("demo.txt","r")
print("First line:")
print(file.readline())

print("Second line:")
print(file.readline())

file.close()

# -------------------------------
print("Append new data")
print("----------------------")
# -------------------------------
file = open("demo.txt", "a")
file.write("This line was appended.\n")
file.close()

# -------------------------------
# 5. Read using loop
# -------------------------------
file = open("demo.txt", "r")

print("\nReading using loop:")
for line in file:
    print(line.strip())

file.close()

# -------------------------------
# 6. Check if file exists
# -------------------------------
if os.path.exists("demo.txt"):
    print("\nFile exists.")
else:
    print("\nFile not found.")

# -------------------------------
# 7. Delete file
# -------------------------------
os.remove("demo.txt")
print("\nFile deleted successfully.")
```
