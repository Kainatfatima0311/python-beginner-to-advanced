## Python OOP (Object-Oriented Programming)

### What is OOP?

- OOP stands for **Object-Oriented Programming**.
- It is a programming style based on **Classes** and **Objects**.
- It helps organize code in a clean and reusable way.

### Class

- A Class is a blueprint or template.
- It defines the attributes and methods of an object.
- Multiple objects can be created from one class.

### Object

- An Object is an instance of a class.
- Every object has its own data.
- Objects use the methods defined inside the class.

### Attributes

- Attributes store data.
- Examples:
  - Name
  - Age
  - Marks

### Methods

- Methods are functions inside a class.
- They define the behavior of an object.

### Why Use OOP?

- Code Reusability
- Better Organization
- Easy Maintenance
- Easier Team Collaboration
- Suitable for Large Projects

### Real-Life Examples

- Car
- Student
- Mobile
- Bank Account

### Important Terms

- Class → Blueprint
- Object → Real Instance
- Attributes → Data
- Methods → Functions

````python
# ==========================================
# Creating a Class
# ==========================================

class Student:
    # This method prints a simple message
    def study(self):
        print("Student is Studying")


# ==========================================
# Creating Objects
# ==========================================

student1 = Student()

student2 = Student()


# ==========================================
# Calling Methods
# ==========================================

student1.study()

student2.study()


print("---------------------")


# ==========================================
# Another Example
# ==========================================

class Car:

    def start(self):
        print("Car Started")

    def stop(self):
        print("Car Stopped")


# Creating Objects
car1 = Car()
car2 = Car()

# Calling Methods
car1.start()
car2.stop()
````

## Python __init__() Function

### What is `__init__()`?

- `__init__()` is a special method.
- It is also called a constructor.
- It runs automatically when an object is created.
- It is used to initialize object data.

### Syntax

```python
class Student:
    def __init__(self):
        pass
```

### self Keyword

- `self` refers to the current object.
- It is used to access object attributes and methods.
- Python passes it automatically.

### Object Attributes

- Attributes store object data.
- They are usually created inside `__init__()`.

Example:

```python
self.name = name
```

### Multiple Parameters

- Constructor can receive multiple values.

Example:

```python
def __init__(self, name, age):
```

### Access Attributes

```python
print(student.name)
```

### Why Use `__init__()`?

- Initialize object data.
- Avoid repeated assignments.
- Make objects ready to use.

### Best Practice

- Initialize important attributes inside `__init__()`.
- Always use `self` to create object attributes.

````python
# ==========================================
# Simple Constructor
# ==========================================

class Student:

    def __init__(self):
        print("Student Object Created")


student1 = Student()


print("----------------------")


# ==========================================
# Constructor with One Parameter
# ==========================================

class Student:

    def __init__(self, name):
        # Store the value inside the object
        self.name = name


student1 = Student("Ali")

print(student1.name)


print("----------------------")


# ==========================================
# Constructor with Multiple Parameters
# ==========================================

class Student:

    def __init__(self, name, age):
        self.name = name
        self.age = age


student1 = Student("Ali", 20)
student2 = Student("Sara", 22)
student3 = Student("Kainat", 21)

print(student1.name, student1.age)

print(student2.name, student2.age)

print(student3.name, student3.age)


print("----------------------")


# ==========================================
# Another Example
# ==========================================

class Car:

    def __init__(self, company, model):
        self.company = company
        self.model = model


car1 = Car("Toyota", "Corolla")

print(car1.company)

print(car1.model)


print("----------------------")


# ==========================================
# Checking Object Data
# ==========================================

print(type(student1))

print(type(car1))
````

## Python self

### What is `self`?

- `self` refers to the current object.
- It is used to access object attributes and methods.
- Python passes it automatically when a method is called.

### Is `self` a Keyword?

- No.
- It is a naming convention.
- Using `self` is the recommended practice.

### Why Use `self`?

- Access object attributes.
- Access object methods.
- Keep each object's data separate.

### Access Attributes

```python
self.name = name
```

### Call Another Method

```python
self.greet()
```

### Important Note

- `self` must be the first parameter of instance methods.
- You do not pass `self` manually.

### Best Practice

- Always use `self` as the first parameter.
- Do not replace it with another name.

````python
# ==========================================
# Using self in __init__()
# ==========================================

class Student:

    def __init__(self, name, age):
        # Store values inside the object
        self.name = name
        self.age = age


student1 = Student("Ali", 20)
student2 = Student("Sara", 22)

print(student1.name)
print(student2.name)


print("----------------------")


# ==========================================
# Using self in Methods
# ==========================================

class Student:

    def __init__(self, name):
        self.name = name

    def introduce(self):
        # Access object attribute using self
        print(f"My name is {self.name}")


student = Student("Kainat")

student.introduce()


print("----------------------")


# ==========================================
# Calling One Method from Another
# ==========================================

class Student:

    def greet(self):
        print("Hello!")

    def welcome(self):
        # Call another method of the same object
        self.greet()
        print("Welcome to Python OOP")


student = Student()

student.welcome()


print("----------------------")


# ==========================================
# Different Objects Have Different Data
# ==========================================

class Car:

    def __init__(self, company):
        self.company = company

    def show_company(self):
        print(self.company)


car1 = Car("Toyota")
car2 = Car("Honda")

car1.show_company()
car2.show_company()
````

## Python Class Properties

### What is a Property?

- A property is a variable that stores data inside a class or object.
- Properties describe an object's information.
- Examples:
  - Name
  - Age
  - City
  - Brand

### Object Properties

- Object properties are also called instance properties.
- They are usually created inside `__init__()`.
- They use `self`.
- Every object can have different values.

Example:

```python
self.name = name
self.age = age
```

### Access Properties

- Use dot notation to access a property.

Syntax:

```python
object.property
```

Example:

```python
student1.name
```

### Modify Properties

- Object properties can be changed after creating the object.

Example:

```python
student1.age = 22
```

### Delete Properties

- Use the `del` keyword to delete an object property.

Example:

```python
del student1.age
```

### Class Properties

- Class properties are defined inside the class but outside methods.
- They are shared by all objects of the class.

Example:

```python
class Student:
    institute = "WorkEdge"
```

### Class Property vs Object Property

| Class Property | Object Property |
|---|---|
| Shared by all objects | Separate for every object |
| Defined outside methods | Usually defined inside `__init__()` |
| Accessed using class or object | Accessed using object |
| Example: institute | Example: name |

### Modify Class Properties

- Use the class name to modify a class property.

Example:

```python
Student.institute = "New Institute"
```

### Add New Object Properties

- New properties can be added after object creation.
- They are added only to that specific object.

Example:

```python
student1.city = "Lahore"
```

### Best Practice

- Define important object properties inside `__init__()`.
- Use class properties for shared data.
- Modify class properties using the class name.

````python
# ==========================================
# Object Properties
# ==========================================

class Student:

    def __init__(self, name, age):
        # These are object properties
        self.name = name
        self.age = age


student1 = Student("Kainat", 21)
student2 = Student("Ali", 20)


# ==========================================
# Access Properties
# ==========================================

# Dot notation is used to access properties
print(student1.name)
print(student1.age)

print(student2.name)
print(student2.age)


print("----------------------")


# ==========================================
# Modify an Object Property
# ==========================================

# Change only student1's age
student1.age = 22

print(student1.age)  # 22
print(student2.age)  # 20


print("----------------------")


# ==========================================
# Delete an Object Property
# ==========================================

del student1.age

print(student1.name)

# This would cause an AttributeError:
# print(student1.age)


print("----------------------")


# ==========================================
# Class Property vs Object Property
# ==========================================

class Employee:

    # Class property
    # Shared by all Employee objects
    company = "WorkEdge"

    def __init__(self, name, role):
        # Object properties
        # Different for every object
        self.name = name
        self.role = role


employee1 = Employee("Kainat", "AI Engineer Intern")
employee2 = Employee("Ali", "Python Developer")

print(employee1.name)
print(employee2.name)

# Both objects use the same class property
print(employee1.company)
print(employee2.company)


print("----------------------")


# ==========================================
# Modify Class Property
# ==========================================

# Change the shared property using class name
Employee.company = "WorkEdge Ltd."

print(employee1.company)
print(employee2.company)


print("----------------------")


# ==========================================
# Add New Property to One Object
# ==========================================

# This property is added only to employee1
employee1.city = "Lahore"

print(employee1.city)

# employee2 does not have a city property
# This would cause an AttributeError:
# print(employee2.city)
````

## Python Inheritance

### What is Inheritance?

- Inheritance allows one class to use the properties and methods of another class.
- It helps reuse existing code.
- It reduces code repetition.

### Parent Class

- The class being inherited from is called the parent class.
- It is also called the base class.

### Child Class

- The class that inherits from another class is called the child class.
- It is also called the derived class.

### Create a Parent Class

- A parent class is created like a normal class.
- It can contain properties and methods.

### Create a Child Class

- Pass the parent class inside the child class brackets.

Syntax:

```python
class ChildClass(ParentClass):
    pass
```

### Inherited Features

- A child class can use the parent's:
  - Properties
  - Methods
  - Constructor

### Child `__init__()`

- A child class can have its own `__init__()` method.
- The child constructor replaces the inherited parent constructor.

### Call Parent `__init__()`

- The parent constructor can be called using the parent class name.

Example:

```python
ParentClass.__init__(self, values)
```

### `super()` Function

- `super()` is used to access the parent class.
- It is commonly used to call the parent constructor.
- `self` is passed automatically.

Example:

```python
super().__init__(values)
```

### Add Child Properties

- A child class can add its own properties.
- These properties are created inside the child constructor.

### Add Child Methods

- A child class can define its own methods.
- It can use both inherited and new methods.

### Benefits of Inheritance

- Code reusability
- Less repetition
- Better organization
- Easier maintenance

### Best Practice

- Use `super()` to call the parent constructor.
- Use inheritance when the child has an “is-a” relationship with the parent.

````python
# ==========================================
# 1. Create a Parent Class
# ==========================================

class Person:

    def __init__(self, name, age):
        # Parent class properties
        self.name = name
        self.age = age

    def introduce(self):
        # Parent class method
        print(f"My name is {self.name}")
        print(f"My age is {self.age}")


person1 = Person("Ali", 25)

person1.introduce()


print("----------------------")


# ==========================================
# 2. Create a Child Class
# ==========================================

class Student(Person):
    # No extra property or method added yet
    pass


# Student uses the inherited Person constructor
student1 = Student("Kainat", 21)

# Student uses the inherited Person method
student1.introduce()


print("----------------------")


# ==========================================
# 3. Child Class with its Own __init__()
# ==========================================

class Employee(Person):

    def __init__(self, name, age, role):
        # Call parent constructor
        Person.__init__(self, name, age)

        # Child class property
        self.role = role

    def show_role(self):
        print(f"My role is {self.role}")


employee1 = Employee("Sara", 24, "Python Developer")

# Inherited method
employee1.introduce()

# Child class method
employee1.show_role()


print("----------------------")


# ==========================================
# 4. Using super()
# ==========================================

class Teacher(Person):

    def __init__(self, name, age, subject):
        # super() calls the parent constructor
        super().__init__(name, age)

        # Child property
        self.subject = subject

    def teach(self):
        print(f"{self.name} teaches {self.subject}")


teacher1 = Teacher("Ahmed", 35, "Python")

teacher1.introduce()
teacher1.teach()


print("----------------------")


# ==========================================
# 5. Parent and Child Properties
# ==========================================

class Student(Person):

    def __init__(self, name, age, graduation_year):
        # Create inherited name and age properties
        super().__init__(name, age)

        # Create child-specific property
        self.graduation_year = graduation_year

    def welcome(self):
        # Access inherited and child properties
        print(
            f"Welcome {self.name}, "
            f"your graduation year is {self.graduation_year}"
        )


student2 = Student("Kainat", 21, 2027)

print(student2.name)
print(student2.age)
print(student2.graduation_year)

student2.introduce()
student2.welcome()
````

## Python Polymorphism

### What is Polymorphism?

- Polymorphism means **many forms**.
- The same function or method name can work differently with different objects.
- It allows one interface to represent different behaviors.

### Simple Example

- Car `move()` → Drive
- Boat `move()` → Sail
- Plane `move()` → Fly

### Function Polymorphism

- One function can work with different data types.
- Python's `len()` function is an example.

Examples:

- String → Number of characters
- Tuple → Number of items
- Dictionary → Number of key-value pairs

### Class Polymorphism

- Different classes can have methods with the same name.
- Each class can define different behavior for that method.

Example:

```python
car.move()
boat.move()
plane.move()
```

### Polymorphism in Loops

- Different objects can be placed in one loop.
- The same method call can be used for every object.

Example:

```python
for vehicle in vehicles:
    vehicle.move()
```

### Inheritance Polymorphism

- Child classes inherit methods from a parent class.
- Child classes can replace inherited behavior.

### Method Overriding

- Method overriding happens when a child class defines a method with the same name as the parent method.
- The child method is used for child objects.

### Benefits

- Reduces repeated conditional code.
- Makes code flexible.
- Makes code easier to extend.
- Allows different objects to be treated similarly.

### Best Practice

- Use the same method name for actions with the same purpose.
- Let each class define its own behavior.
- Use inheritance only when classes have a clear relationship.

````python
# ==========================================
# 1. Function Polymorphism
# ==========================================

name = "Kainat"

fruits = ("Apple", "Mango", "Banana")

student = {
    "name": "Kainat",
    "age": 21
}


# Same len() function works differently
# according to the provided data type

print(len(name))       # Number of characters
print(len(fruits))     # Number of tuple items
print(len(student))    # Number of key-value pairs


print("----------------------")


# ==========================================
# 2. Class Polymorphism
# ==========================================

class Car:

    def __init__(self, brand):
        self.brand = brand

    def move(self):
        print(f"{self.brand} car drives")


class Boat:

    def __init__(self, brand):
        self.brand = brand

    def move(self):
        print(f"{self.brand} boat sails")


class Plane:

    def __init__(self, brand):
        self.brand = brand

    def move(self):
        print(f"{self.brand} plane flies")


car1 = Car("Toyota")
boat1 = Boat("Yamaha")
plane1 = Plane("Boeing")


# Same method name, different behavior
car1.move()
boat1.move()
plane1.move()


print("----------------------")


# ==========================================
# 3. Polymorphism in a Loop
# ==========================================

vehicles = (car1, boat1, plane1)


for vehicle in vehicles:

    # Python calls the move() method
    # belonging to the current object

    vehicle.move()


print("----------------------")


# ==========================================
# 4. Polymorphism with Inheritance
# ==========================================

class Vehicle:

    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

    def move(self):
        # Default parent behavior
        print("Vehicle moves")


class Car(Vehicle):

    # Car does not create its own move()
    # It will use Vehicle's move()
    pass


class Boat(Vehicle):

    # This method overrides Vehicle.move()
    def move(self):
        print("Boat sails")


class Plane(Vehicle):

    # This method overrides Vehicle.move()
    def move(self):
        print("Plane flies")


car2 = Car("Toyota", "Corolla")
boat2 = Boat("Yamaha", "242X")
plane2 = Plane("Boeing", "747")


vehicles = (car2, boat2, plane2)


for vehicle in vehicles:

    print("Brand:", vehicle.brand)
    print("Model:", vehicle.model)

    # Same method call gives different output
    vehicle.move()

    print()
````

## Python Encapsulation

### What is Encapsulation?

- Encapsulation means bundling data and methods inside a class.
- It helps protect object data.
- It provides controlled access to data.

### Public Members

- Public members are accessible from anywhere.
- They are created normally.

Example:

```python
self.name = name
```

### Protected Members

- Protected members start with a single underscore `_`.
- They are intended for internal use.
- Python does not strictly prevent access.

Example:

```python
self._name = name
```

### Private Members

- Private members start with a double underscore `__`.
- They are hidden using name mangling.
- They should be accessed through class methods.

Example:

```python
self.__password = password
```

### Getter Method

- A getter returns the value of a private member.

Example:

```python
def get_balance(self):
    return self.__balance
```

### Setter Method

- A setter updates the value of a private member.

Example:

```python
def set_balance(self, amount):
    self.__balance = amount
```

### Benefits of Encapsulation

- Protects object data.
- Prevents unwanted changes.
- Allows validation.
- Improves code security and maintainability.

### Best Practice

- Use public members for general data.
- Use protected members for internal implementation.
- Use private members for sensitive data.
- Access private members through methods.

````python
# ==========================================
# 1. Public Member
# ==========================================

class Student:

    def __init__(self, name):
        self.name = name


student = Student("Kainat")

print(student.name)

student.name = "Ali"

print(student.name)


print("----------------------")


# ==========================================
# 2. Protected Member
# ==========================================

class Student:

    def __init__(self, name):
        self._name = name


student = Student("Kainat")

# Accessible, but intended for internal use
print(student._name)


print("----------------------")


# ==========================================
# 3. Private Member
# ==========================================

class Student:

    def __init__(self):
        self.__password = "12345"

    def show_password(self):
        print(self.__password)


student = Student()

# Correct way
student.show_password()

# Wrong (will raise AttributeError)
# print(student.__password)


print("----------------------")


# ==========================================
# 4. Getter and Setter
# ==========================================

class BankAccount:

    def __init__(self):
        self.__balance = 10000

    def get_balance(self):
        return self.__balance

    def set_balance(self, amount):

        # Validation
        if amount >= 0:
            self.__balance = amount
        else:
            print("Balance cannot be negative")


account = BankAccount()

print(account.get_balance())

account.set_balance(5000)

print(account.get_balance())

account.set_balance(-100)


print("----------------------")


# ==========================================
# 5. Real Example
# ==========================================

class User:

    def __init__(self, username, password):

        self.username = username      # Public
        self.__password = password    # Private

    def change_password(self, new_password):
        self.__password = new_password

    def show_user(self):
        print("Username:", self.username)
        print("Password:", self.__password)


user = User("Kainat", "abc123")

user.show_user()

user.change_password("Python123")

user.show_user()
````

## Python Inner Classes

### What is an Inner Class?

- An inner class is a class defined inside another class.
- The containing class is called the outer class.
- Inner classes help group closely related classes.
- They make code more organized.

### Basic Syntax

```python
class Outer:

    class Inner:
        pass
```

### Create an Inner Class Object

- First create the outer class object.
- Access the inner class through the outer object.
- Create the inner class object using parentheses.

Example:

```python
outer = Outer()
inner = outer.Inner()
```

### Separate Objects

- Outer and inner classes create separate objects.
- Each object has its own `self`.
- Both classes can have their own properties and methods.

### Access Outer Data

- An inner class does not automatically receive the outer object.
- Pass the outer object as a parameter when required.

Example:

```python
inner = outer.Inner(outer)
```

### Store Inner Object

- An outer object can create and store an inner object.

Example:

```python
self.engine = self.Engine()
```

### Access Stored Inner Object

Example:

```python
car.engine.start()
```

### Multiple Inner Classes

- One outer class can contain multiple inner classes.

Examples:

- Computer → CPU
- Computer → RAM

### Uses of Inner Classes

- Group related classes.
- Create helper classes.
- Represent internal parts of an object.
- Improve code organization.

### Best Practice

- Use an inner class when it mainly belongs to one outer class.
- Use a separate class when it is required independently in many places.

````python
# ==========================================
# 1. Simple Inner Class
# ==========================================

class Outer:

    def __init__(self):
        self.name = "Outer Class"

    class Inner:

        def __init__(self):
            self.name = "Inner Class"

        def display(self):
            print("Hello from inner class")


# Create outer object
outer = Outer()

# Create inner object through outer class
inner = outer.Inner()

print(outer.name)
print(inner.name)

inner.display()


print("----------------------")


# ==========================================
# 2. Access Outer Data from Inner Class
# ==========================================

class University:

    def __init__(self, university_name):
        self.university_name = university_name

    class Department:

        def __init__(self, outer, department_name):
            # Save the outer University object
            self.outer = outer

            # Save inner class property
            self.department_name = department_name

        def show_details(self):
            # Access outer and inner properties
            print("University:", self.outer.university_name)
            print("Department:", self.department_name)


university = University("ABC University")

# Pass outer object to the inner class
department = university.Department(
    university,
    "Computer Science"
)

department.show_details()


print("----------------------")


# ==========================================
# 3. Practical Car and Engine Example
# ==========================================

class Car:

    def __init__(self, brand, model):
        self.brand = brand
        self.model = model

        # Create and store Engine object
        self.engine = self.Engine()

    class Engine:

        def __init__(self):
            self.status = "Off"

        def start(self):
            self.status = "Running"
            print("Engine started")

        def stop(self):
            self.status = "Off"
            print("Engine stopped")

    def drive(self):

        # Check the inner Engine object's property
        if self.engine.status == "Running":
            print(f"Driving the {self.brand} {self.model}")
        else:
            print("Start the engine first")


car = Car("Toyota", "Corolla")

# Engine is currently off
car.drive()

# Access inner object and call its method
car.engine.start()

# Now the car can drive
car.drive()

car.engine.stop()


print("----------------------")


# ==========================================
# 4. Multiple Inner Classes
# ==========================================

class Computer:

    def __init__(self):
        # Create objects of both inner classes
        self.cpu = self.CPU()
        self.ram = self.RAM()

    class CPU:

        def process(self):
            print("CPU is processing data")

    class RAM:

        def store(self):
            print("RAM is storing data")


computer = Computer()

computer.cpu.process()
computer.ram.store()
````
