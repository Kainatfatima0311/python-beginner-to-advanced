#### **Python String Slicing**

Slicing is used to get a **part of a string**.

- Use the syntax `string[start:end]`.
- The **start index is included**.
- The **end index is not included**.

````python
text = "Python"

print(text[0:2])
````

````python
text = "Python"

print(text[:4])
````

````python
b = "Hello, World!"
print(b[2:])
````

````python
text = "Python"

print(text[0:6:2])
````

````python
txt ="python"
print(txt[:])
````

#### **Operator Precedence**

Operator precedence decides **which operation is performed first** when an expression contains multiple operators.

| **Priority** | **Operator** |
|--------------|--------------|
| 1 | `()` |
| 2 | `**` |
| 3 | `*`, `/`, `//`, `%` |
| 4 | `+`, `-` |
| 5 | `==`, `!=`, `>`, `<`, `>=`, `<=` |
| 6 | `not` |
| 7 | `and` |
| 8 | `or` |

````python
# ==========================
# Arithmetic Operators
# ==========================

a = 10
b = 5

# Addition
print(a + b)

# Subtraction
print(a - b)

# Multiplication
print(a * b)

# Division
print(a / b)

# Floor Division (Removes the decimal part)
print(a // b)

# Modulus (Returns the remainder)
print(a % b)

# Exponent (Power)
print(a ** b)


# ==========================
# Assignment Operators
# ==========================

x = 10
print("Initial Value:", x)

x += 5    # x = x + 5
print("After += :", x)

x -= 10   # x = x - 10
print("After -= :", x)

x *= 20   # x = x * 20
print("After *= :", x)


# ==========================
# Comparison Operators
# ==========================

c = 10
d = 7

print(c == d)   # Equal
print(c != d)   # Not Equal
print(c > d)    # Greater Than
print(c < d)    # Less Than
print(c >= d)   # Greater Than or Equal
print(c <= d)   # Less Than or Equal


# ==========================
# Logical Operators
# ==========================

# AND Operator (Both conditions must be True)
age = 19
student = True

print(age >= 18 and student)

# OR Operator (At least one condition must be True)
print(age >= 18 or student)

# NOT Operator (Reverses the result)
is_login = True

print(not is_login)
````

````python
# ==========================
# Identity Operators
# ==========================
# Identity operators compare the memory location (object identity).

a = [1, 2, 3]
b = a

print(a is b)      # True (Both refer to the same object)

b = [1, 2, 3]

print(a == b)      # True (Values are the same)
print(a is b)      # False (Different objects in memory)


# ==========================
# Membership Operators
# ==========================
# Check whether a value exists in a collection.

fruits = ["Apple", "Mango", "Banana"]

print("Mango" in fruits)        # True
print("Orange" not in fruits)   # True


# ==========================
# Bitwise Operators
# ==========================
# Bitwise operators work on binary (0 and 1) values.

a = 5   # Binary: 101
b = 3   # Binary: 011

print(a & b)   # AND
print(a | b)   # OR
print(a ^ b)   # XOR
print(~a)      # NOT
print(a << 1)  # Left Shift
print(a >> 1)  # Right Shift


# ==========================
# Ternary Operator
# ==========================
# value_if_true if condition else value_if_false

age = 21

result = "Adult" if age >= 18 else "Not Adult"

print(result)
````

#### remove()

- Removes an item by value.
- Removes only the first matching item.
- Raises `ValueError` if the value is not found.
- Does not use an index.

**Syntax:**

list_name.remove(value)

#### pop()

- Removes an item by index.
- By default, removes the last item.
- Returns the removed item.
- Raises `IndexError` if the index is out of range.

**Syntax:**

```python
list_name.pop(index)
```

Or

```python
list_name.pop()
```

````python
#remove item
colors=["red","orange","blue"]
colors.remove("blue")
print(colors)

#===================
marks=[10,20,57,78,65,87,90]
marks.pop(2)
print(marks)
marks.pop()
print(marks)
marks.clear()
print(marks)
````

#### Loop Through a List

- Used to iterate through all items in a list.
- `for` loop accesses one item at a time.
- The loop automatically stops after the last item.

**Syntax:**

```python
for item in list_name:
    print(item)
```

````python
fruits=["apple","mango","orange","berry"]
for i in fruits:
  print(i)
````

# Loop Using Index

- Uses `range()` and `len()`.
- Gives access to index numbers.

```python
for i in range(len(list_name)):
    print(list_name[i])
```

````python
fruits = ["Apple", "Banana", "Mango"]

for i in range(len(fruits)):
    print(i, fruits[i])
````

# while Loop

- Repeats while the condition is `True`.
- Index must be updated manually.

```python
i = 0
while i < len(list_name):
    print(list_name[i])
    i += 1
```

````python
fruits=["apple","mango","orange"]
i=0
while i<len(fruits):
  print(fruits[i])
  i+=1
````

# List Comprehension Loop

- Can be used for short loops.
- Mostly used to create new lists.

```python
[print(item) for item in list_name]
```

````python
marks=[90,87,68,58,47]
[print(mark) for mark in marks]

print("=============")
fruits = ["Apple", "Banana", "Mango"]

[print(fruit) for fruit in fruits]
````
