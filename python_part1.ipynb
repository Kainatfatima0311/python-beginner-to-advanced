**What is Python?**

Python is a **high-level**, **easy-to-learn**, and **popular** programming language.

- Created by **Guido van Rossum** in **1991**.
- Used for **Web Development, Software Development, AI, Automation, Data Science,** and **Mathematics**.
- Has a **simple and readable syntax**, making it a great choice for beginners.
- An **interpreted language**, so code is executed line by line.
- Uses **indentation** instead of curly braces `{}` to define code blocks.
- Supports **Procedural**, **Object-Oriented (OOP)**, and **Functional Programming**.

**Why Python?**

- **Simple Design** → Python is easy to read, write, and understand.
- **Batteries Included** → Python has a large Standard Library, so you don't need to write everything from scratch. Many tasks can be done with just a few lines of code.
- **General-Purpose Language** → Python can be used for Web Development, Data Science, AI, Machine Learning, Automation, Game Development, and more.
- **Rich Libraries** → Python already has thousands of built-in and third-party libraries that make coding faster and easier.


#### **Why Python for Data Science?**

- **Easy to Learn** → Python has simple syntax, so beginners can learn it quickly.
- **Powerful Libraries** → Libraries like **NumPy**, **Pandas**, **Matplotlib**, and **Scikit-learn** make data analysis and machine learning easier.
- **Strong Community** → Python has a large community. If you get stuck, you can easily find tutorials, documentation, and solutions online.

#### Python Variables

A **variable** is a container that stores data.

- In Python, a variable is created when you assign a value to it.
- You do **not** need to specify the data type. Python automatically detects it.
- Variables can store different types of data, such as numbers, text, and boolean values.

**Syntax:**

```python
variable_name = value
```

**Multi Words Variable Names**

**Camel Case**
Each word, except the first, starts with a capital letter

**Pascal Case**
Each word starts with a capital letter

**Snake Case**
Each word is separated by an underscore character

**Note:** **Snake Case** is the recommended naming style in Python (PEP 8).

````python
myName="kainat" # Camel Case
MyName="fatima" # Pascal Case
my_name="kf" # Snake Case (Recommended in Python)
````

````python
name="kainat"
x=5
age="21 years"

print(name)
print(x)
print(age)
````

#### **Variable Naming Rules**

- A variable name **must start with a letter** (`a-z`, `A-Z`).
- It can also **start with an underscore (`_`)**.
- A **number cannot be the first character**, but it can come later.
- **Spaces are not allowed** in variable names.
- **Special characters** (such as `@`, `#`, `$`, `%`) are **not allowed** except the underscore (`_`).
- **Python keywords** (such as `if`, `for`, `while`, `class`) **cannot** be used as variable names.

#### **Important Points**

##### **Variable Value Can Change**
- A variable's value can be updated at any time.

##### **Python is Case Sensitive**
- Variable names with different letter cases are treated as different variables.

##### **Multiple Variables in One Line**
- You can assign values to multiple variables in a single line.

##### **Same Value to Multiple Variables**
- You can assign the same value to multiple variables at once.

##### **Single and Double Quotes**
- You can use **single (`' '`)** or **double (`" "`)** quotes for strings. Both work the same way.

####**Many Values to Multiple Variables**

````python
x,y,z=10,20,30

print(x)
print(y)
print(z)
````

####**One Value to Multiple Variables**

````python
x=y=z=10

print(x)
print(y)
print(z)
````
