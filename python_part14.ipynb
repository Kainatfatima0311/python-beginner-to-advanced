## NumPy Random Intro

### What is Random?

- Random means unpredictable values.
- Each execution may produce different results.

### NumPy Random Module

```python
np.random
```

### Random Integer

```python
np.random.randint(start, end)
```

- Generates a random integer.
- Start is included.
- End is excluded.

### Random Float

```python
np.random.rand()
```

- Generates a random decimal number between 0 and 1.

### Uses

- Games
- OTP Generation
- Machine Learning
- Statistics
- Simulations

### Best Practice

- Remember that random values change every time you run the program.

````python
import numpy as np

num = np.random.randint(1, 10)

print(num)
````

## NumPy Random Data Distribution

### What is Data Distribution?

- Data distribution defines the probability of selecting each value.
- Some values can appear more often than others.

### `np.random.choice()`

```python
np.random.choice(values, p=probabilities)
```

### Parameters

- `values` → Values to choose from.
- `p` → Probability of each value.

### Example

```python
np.random.choice(
    [1,2,3],
    p=[0.6,0.3,0.1]
)
```

### Multiple Values

```python
np.random.choice(
    [1,2,3],
    size=5,
    p=[0.6,0.3,0.1]
)
```

### Important Rule

- The sum of all probabilities must be **1**.

### Best Practice

- Use `choice()` when different values should have different chances of being selected.

````python
import numpy as np

# ==========================================
# Random Value with Probability
# ==========================================

print("Single Random Value:")
print(
    np.random.choice(
        [1, 2, 3],
        p=[0.6, 0.3, 0.1]
    )
)

print("------------------")

# ==========================================
# Multiple Random Values
# ==========================================

print("Five Random Values:")
print(
    np.random.choice(
        [1, 2, 3],
        size=5,
        p=[0.6, 0.3, 0.1]
    )
)
````

## NumPy Random Permutation

### What is Permutation?

- Permutation means changing the order of elements.
- Values remain the same, only their positions change.

### `shuffle()`

```python
np.random.shuffle(arr)
```

- Randomly rearranges the array.
- Changes the original array.

### `permutation()`

```python
np.random.permutation(arr)
```

- Returns a new shuffled array.
- Does not change the original array.

### Difference

| shuffle() | permutation() |
|------------|---------------|
| Changes original array | Returns a new array |
| No new array | Original remains unchanged |

### Best Practice

- Use `shuffle()` when you want to modify the original array.
- Use `permutation()` when you want to keep the original array safe.

````python
import numpy as np

# ==========================================
# shuffle()
# ==========================================

arr1 = np.array([1, 2, 3, 4, 5])

np.random.shuffle(arr1)

print("After shuffle():")
print(arr1)

print("------------------")

# ==========================================
# permutation()
# ==========================================

arr2 = np.array([1, 2, 3, 4, 5])

new_arr = np.random.permutation(arr2)

print("New Permutation:")
print(new_arr)

print("Original Array:")
print(arr2)
````

## NumPy Random Normal Distribution

### What is Normal Distribution?

- Normal Distribution is a bell-shaped distribution.
- Most values are near the mean.
- Fewer values appear at the extremes.

### Function

```python
np.random.normal(loc, scale, size)
```

### Parameters

- `loc` → Mean (center)
- `scale` → Standard deviation (spread)
- `size` → Number of values

### Example

```python
np.random.normal(
    loc=50,
    scale=5,
    size=10
)
```

### Graph

```python
sns.displot(data)
```

### Uses

- Student Marks
- Height
- Weight
- IQ Scores
- Machine Learning

### Best Practice

- `loc` controls the center.
- `scale` controls the spread.
- `size` controls the number of generated values.

````python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# ==========================================
# Generate Normal Distribution Data
# ==========================================

data = np.random.normal(
    loc=50,
    scale=5,
    size=1000
)

print("First 10 Values:")
print(data[:10])

# ==========================================
# Plot Distribution
# ==========================================

sns.displot(data)

plt.show()
````

## NumPy Random Multinomial Distribution

### What is Multinomial Distribution?

- Used when there are more than two possible outcomes.
- It is an extension of the Binomial Distribution.

### Function

```python
np.random.multinomial(n, pvals, size)
```

### Parameters

- `n` → Total number of trials
- `pvals` → Probability of each outcome
- `size` → Number of experiments

### Example

```python
np.random.multinomial(
    n=10,
    pvals=[0.2,0.5,0.3],
    size=1
)
```

### Uses

- Election Results
- Dice Rolls
- Surveys
- Customer Preferences

### Best Practice

- The sum of all probabilities in `pvals` must be **1**.
- The sum of values in each result will always equal `n`.

````python
import numpy as np

# ==========================================
# Generate Multinomial Distribution
# ==========================================

data = np.random.multinomial(
    n=10,
    pvals=[0.2, 0.5, 0.3],
    size=3
)

print("Results:")
print(data)
````
